# Метод наименьших квадратов (Least squares approximation)

**Цель работы:** ознакомление с методом аппроксимации функций.

**Вариант:** 14

## Задание

Дана таблица значений некоторой функциональной зависимости, полученной из $n = 6$ опытов

1. Методом наименьших квадратов по данной табличной зависимости найти аппроксимирующую функцию в виде:

  1.1) линейной функции $y = ax + b$;

  1.2) степенной функции $y = \beta \cdot x^a$;

  1.3) показательной функции $y = \beta \cdot e^{ax}$;

  1.4) квадратичной функции $y = ax^2 + bx + c$.

2. Построить в плоскости $xOy$ графики полученных функций и нанести экспериментальные точки.

3. Сравнить полученные результаты.

## Решение

|$x$ | 5 | 15 | 25 | 35 | 45 | 55 |

**1.1  Найдем зависимость $y$ от $x$ в виде линейной функции $y = ax + b$.**

Выберем значения коэффициентов $a$ и $b$ так, чтобы сумма квадратов отклонений
$ L(a, b) = \sum\limits_{i=1}^n (\Delta y_i)^2 = 
\sum\limits_{i=1}^{n} { \left((ax_i+b)-y_i\right)^2}$
была минимальной.

Имеем, 

$\sum\limits_{i=1}^{6} {x_i} = 5 + 15 + 25+ 35 + 45 + 55 = 180$,

$\sum\limits_{i=1}^{6} {y_i} =  2,2 + 2,4 + 2,6 + 2,7 + 2,9 = 15,6$,

$\sum\limits_{i=1}^{6} {x_i^2} = 5^2 + 15^2 + 25^2 + 35^2 + 45^2 + 55^2 = 7150$,

$\sum\limits_{i=1}^{6} {(x_i \cdot y_i)} = 
5\cdot 2,2 + 15\cdot 2,4 + 25\cdot 2,6 + 35\cdot 2,7 + 45\cdot 2,8 + 55\cdot 2,9 = 492$.

Тогда система уравнений (7) примет вид:

\begin{equation}
    \begin{cases}
        7150a + 180b = 492, \\
        180a + 6b = 15,6.
    \end{cases}
\end{equation}

Решим систему уравнений по формулам Крамера:

\begin{equation}
    a = \cfrac{\Delta_a}{\Delta}, \;
    b = \cfrac{\Delta_b}{\Delta},
\end{equation}
где
\begin{equation}
    \Delta =
    \begin{vmatrix}
        7150 & 180 \\
        180 & 6
    \end{vmatrix} = 7150 \cdot 6 - 180^2 = 10500, 
\end{equation}
\begin{equation}
    \Delta_a =
    \begin{vmatrix}
        492 & 180 \\
        15,6 & 6
    \end{vmatrix} = 492 \cdot 6 - 180 \cdot 15,6 = 144, 
    \quad
    \Delta_b =
    \begin{vmatrix}
        7150 & 492 \\
        180 & 15,6 
    \end{vmatrix} = 7150 \cdot 15,6 - 492 \cdot 180 = 22980. 
\end{equation}

Тогда,
\begin{equation}
    a = \cfrac{\Delta_a}{\Delta} = \cfrac{144}{10500} = 0,0137, 
    \quad
    b = \cfrac{\Delta_b}{\Delta} = \cfrac{22980}{10500} = 2,189.
\end{equation}

Следовательно, искомая линейная функция будет иметь вид:
\begin{equation}
    y = 0,01 x + 2,19.
\end{equation}



**1.2  Найдем зависимость $y$ от $x$ в виде степенной функции $y = \beta \cdot x^a$.**

Прологарифмировав равенство $y = \beta \cdot x^a$ по основанию $e$, получим
$\ln{y} = a\cdot \ln{x} + \ln{\beta}$.

Обозначим
$Y = \ln{y}$, $X = \ln{x}$, $b = \ln{\beta}$.

Тогда получим линейную функцию $Y = a \cdot X + b$, где переменные $X$ и $Y$ связаны следующей табличной зависимостью:

|$X = \ln{x}$ |1,6094 | 2,7081 | 3,2189 | 3,5553 | 3,8067 | 4,0073 |
|:------------|:---:|:---:|:---:|:---:|:---:|:---:|
|$Y = \ln{y}$ |  0,7885 | 0,8755 | 0,9555 | 0,9933 | 1,0296 | 1,0647 |

Таким образом, данная задача свелась к задаче 1.1.

Имеем,

$\sum\limits_{i=1}^{6} {\ln{x_i}} = 1,6094 + 2,7081 + 3,2189 + 3,5553 + 3,8067 + 4,0073 = 18,9057$,

$\sum\limits_{i=1}^{6} {\ln{y_i}} =   0,7885 + 0,8755 + 0,9555 + 0,9933 + 1,0296 + 1,0647 = 5,7071$,

$\sum\limits_{i=1}^{6} {\ln^2{x_i}} = 1,6094^2 + 2,7081^2 + 3,2189^2 + 3,5553^2 + 3,8067^2 + 4,0073^2 = 63,47486745 $,

$\sum\limits_{i=1}^{6} {(\ln{x_i} \cdot \ln{y_i})}
 = 1,6094 \cdot 0,7885 + 2,7081 \cdot 0,8755 + 3,2189 \cdot 0,9555 + 3,5553 \cdot 0,9933 + 3,8067 \cdot 1,0296 + 4,0073 \cdot
 1,0647 = $ $ = 18,43304252$.

Система уравнений (7) будет иметь вид:

\begin{equation}
    \begin{cases}
        63,4749 a + 18,9057 b = 18,43304, \\
        18,9057 a + 6b = 5,7071.
    \end{cases}
\end{equation}

Определители системы:
\begin{equation}
    \Delta =
    \begin{vmatrix}
        63,4749 & 18,9057 \\
        18,9057 & 6
    \end{vmatrix} = 63,4749 \cdot 6 - 18,9057 \cdot 18,9057 = 23,4239,
\end{equation}
\begin{equation}
    \Delta_a =
    \begin{vmatrix}
        18,43304 & 18,9057 \\
        5,7071 & 6
    \end{vmatrix} = 18,43304 \cdot 6 - 18,9057  \cdot 5,7071 = 2,7015, 
\end{equation}    
\begin{equation} 
    \quad    
    \Delta_b =
    \begin{vmatrix}
        63,4749 &  18,43304 \\
        18,9057 &  5,7071 
    \end{vmatrix} = 63,4749 \cdot 5,7071 - 18,43304  \cdot 18,9057 = 13,7681. 
\end{equation}

Тогда,
\begin{equation}
    a = \cfrac{\Delta_a}{\Delta} = \cfrac{2,7015}{23,4239} = 0,1153, 
    \quad
    b = \cfrac{\Delta_b}{\Delta} = \cfrac{13,7681}{23,4239} = 0,5878.
\end{equation}

Учитывая, что $b = \ln{\beta}$, находим $\beta = e^b = e^{0,5878} = 1,80$, 
и получаем искомую степенную функцию 
\begin{equation}
    y = 1,80 \cdot x^{0,12}.
\end{equation}

**1.3  Найдем зависимость $y$ от $x$ в виде показательной функции $y = \beta \cdot e^{ax}$.**


Прологарифмировав равенство $y = \beta \cdot e^{ax}$ по основанию $e$, получим
$\ln{y} = ax  + \ln{\beta}$.

Обозначим
$Y = \ln{y}$, $b = \ln{\beta}$.

Тогда получим линейную функцию $Y = a \cdot x + b$, где переменные $x$ и $Y$ связаны следующей табличной зависимостью:

|$x         $ | 5 | 15 | 25 | 35 | 45 | 55 |
|:------------|:---:|:---:|:---:|:---:|:---:|:---:|
|$Y = \ln{y}$ |  0,7885 | 0,8755 | 0,9555 | 0,9933 | 1,0296 | 1,0647 |

Таким образом, данная задача свелась к задаче 1.1.

Имеем,

$\sum\limits_{i=1}^{6} {x_i} = 5 + 15 + 25+ 35 + 45 + 55 = 180$,

$\sum\limits_{i=1}^{6} {x_i^2} = 5^2 + 15^2 + 25^2 + 35^2 + 45^2 + 55^2 = 7150$,

$\sum\limits_{i=1}^{6} {\ln{y_i}} =  0,7885 + 0,8755 + 0,9555 + 0,9933 + 1,0296 + 1,0647 = 5,7071$,

$\sum\limits_{i=1}^{6} {(x_i \cdot \ln{y_i})} = 5 \cdot 0,7885 + 15 \cdot 0,8755 + 25 \cdot 0,9555 + 35 \cdot 0,9933 + 45 \cdot 1,0296 + 55 \cdot 1,0647$ = 180.6185.

Система уравнений (7) будет иметь вид:

\begin{equation}
    \begin{cases}
        7150 a + 180 b = 180.6185, \\
        180 a + 6b = 5,7071.
    \end{cases}
\end{equation}

Определители системы:
\begin{equation}
    \Delta =
    \begin{vmatrix}
        7150  & 180 \\
        180 & 6
    \end{vmatrix} = 7150 \cdot 6 - 180 \cdot 180 = 10500,
\end{equation}
\begin{equation}
    \Delta_a =
    \begin{vmatrix}
        180,6185 & 180 \\
        5,7071 & 6
    \end{vmatrix} = 180,6185 \cdot 6 - 180  \cdot 5,7071 = 56,433, 
\end{equation}    
\begin{equation}   
    \quad
    \Delta_b =
    \begin{vmatrix}
        7150 &  180,6185 \\
        180 &  5,7071 
    \end{vmatrix} = 7150 \cdot 5,7071 - 180,6185 \cdot 180 = 8294,435. 
\end{equation}

Тогда,
\begin{equation}
    a = \cfrac{\Delta_a}{\Delta} = \cfrac{56,433}{10500} = 0,00537, 
    \quad
    b = \cfrac{\Delta_b}{\Delta} = \cfrac{8294,435}{10500} = 0,7899.
\end{equation}

Учитывая, что $b = \ln{\beta}$, находим $\beta = e^b = e^{0,7899} = 2,2032$, 
и получаем искомую степенную функцию 
\begin{equation}
    y = 2,20 \cdot e^{0,01 x}.
\end{equation}

**1.4 Найдем зависимость $y$ от $x$ в виде показательной функции $y = ax^2 + bx + c$.**

Имеем,

$\sum\limits_{i=1}^{6} {x_i} = 5 + 15 + 25+ 35 + 45 + 55 = 180$,

$\sum\limits_{i=1}^{6} {x_i^2} = 5^2 + 15^2 + 25^2 + 35^2 + 45^2 + 55^2 = 7150$,

$\sum\limits_{i=1}^{6} {x_i^3} = 5^3 + 15^3 + 25^3 + 35^3 + 45^3 + 55^3 = 319500$,

$\sum\limits_{i=1}^{6} {x_i^4} = 5^4 + 15^4 + 25^4 + 35^4 + 45^4 + 55^4 = 15193750$,

$\sum\limits_{i=1}^{6} {y_i} =  2,2 + 2,4 + 2,6 + 2,7 + 2,8 + 2,9 = 15,6$,

$\sum\limits_{i=1}^{6} {(x_i \cdot y_i)} = 
5\cdot 2,2 + 15\cdot 2,4 + 25\cdot 2,6 + 35\cdot 2,7 + 45\cdot 2,8 + 55\cdot 2,9 = 492$.

$\sum\limits_{i=1}^{6} {(x_i^2 \cdot y_i)} = 
5^2 \cdot 2,2 + 15^2 \cdot 2,4 + 25^2 \cdot 2,6 + 35^2 \cdot 2,7 + 45^2 \cdot 2,8 + 55^2 \cdot 2,9 = 19970$.

Система уравнений (15) будет иметь вид:

        319500 a + 7150 b + 180 c = 492,\\
        7150 a + 180 b + 6 c = 15,6.

Решим систему уравенией методом Крамера, учитывая что:

$\Delta = 3920000000$, 
$\Delta_a = -630000$, 
$\Delta_b = 91560000$, 
$\Delta_c = 8195950000$.

Получим,
\begin{equation}
    a = \cfrac{\Delta_a}{\Delta} = \cfrac{-630000}{3920000000} = -0,00016, 
    \quad
    b = \cfrac{\Delta_b}{\Delta} = \cfrac{91560000}{3920000000} = 0.0234,
    \quad
    с = \cfrac{\Delta_с}{\Delta} = \cfrac{8195950000}{3920000000} = 2.0908.
\end{equation}

Следовательно, искомая квадратичная функция будет иметь вид:
\begin{equation}
