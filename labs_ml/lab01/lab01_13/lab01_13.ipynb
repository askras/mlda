1.1  Найдем зависимость $y$ от $x$ в виде линейной функции $y = ax + b$.

Выберем значения коэффициентов $a$ и $b$ так, чтобы сумма квадратов отклонений
$ L(a, b) = \sum\limits_{i=1}^n (\Delta y_i)^2 = 
\sum\limits_{i=1}^{n} { \left((ax_i+b)-y_i\right)^2}$
была минимальной.

Имеем, 

$\sum\limits_{i=1}^{6} {x_i} = 1 + 16 + 13 + 46 + 61 + 76 = 213$,

$\sum\limits_{i=1}^{6} {y_i} =  0,5 + 4,0 + 6,9 + 8,8 + 10,9 + 12,1 = 43,2$,

$\sum\limits_{i=1}^{6} {x_i^2} = 1^2 + 16^2 + 13^2 + 46^2 + 61^2 + 76^2 = 1 + 256 + 169 + 2116 + 3721 + 5776 = 12039$,

$\sum\limits_{i=1}^{6} {(x_i \cdot y_i)} = 
1\cdot 0,5 + 16\cdot 4,0 + 13\cdot 6,9 + 46\cdot 8,8 + 61\cdot 10,9 + 76\cdot 12,1 = 
0,5 + 64,0 + 89,7 + 404,8 + 664,9 + 919,6 = 2143,5$.

Тогда система уравнений примет вид:

\begin{equation}
    \begin{cases}
        12039a + 213b = 2143,5, \\
        213a + 6b = 43,2.
    \end{cases}
\end{equation}

Решим систему уравнений по формулам Крамера:

\begin{equation}
    a = \cfrac{\Delta_a}{\Delta}, \;
    b = \cfrac{\Delta_b}{\Delta},
\end{equation}
где
\begin{equation}
    \Delta =
    \begin{vmatrix}
        12039 & 213 \\
        213 & 6
    \end{vmatrix} = 12039 \cdot 6 - 213 \cdot 213 = 26865, 
\end{equation}
\begin{equation}
    \Delta_a =
    \begin{vmatrix}
        2143,5 & 213 \\
        43,2 & 6
    \end{vmatrix} = 2143,5 \cdot 6 - 213 \cdot 43,2 = 3657,6, 
    \quad
    \Delta_b =
    \begin{vmatrix}
        12039 & 2143,5 \\
        213 & 43,2 
    \end{vmatrix} = 12039 \cdot 43,2 - 2143,5 \cdot 213 = 63519,3. 
\end{equation}

Тогда,
\begin{equation}
    a = \cfrac{\Delta_a}{\Delta} = \cfrac{3657,6}{26865} = 0,136147, 
    \quad
    b = \cfrac{\Delta_b}{\Delta} = \cfrac{63519,3}{26865} = 2,36439.
\end{equation}

Следовательно, искомая линейная функция будет иметь вид:
\begin{equation}
    y = 0,136147 x + 2,36439.
\end{equation}

1.2  Найдем зависимость $y$ от $x$ в виде степенной функции $y = \beta \cdot x^a$.

Прологарифмировав равенство $y = \beta \cdot x^a$ по основанию $e$, получим
$\ln{y} = a\cdot \ln{x} + \ln{\beta}$.

Обозначим
$Y = \ln{y}$, $X = \ln{x}$, $b = \ln{\beta}$.

Тогда получим линейную функцию $Y = a \cdot X + b$, где переменные $X$ и $Y$ связаны следующей табличной зависимостью:

|$x$ | 1 | 16 | 13 | 46 | 61 | 76 | 
|:--:|:---:|:---:|:---:|:---:|:---:|:---:|
|$y$ | 0,5 | 4,0 | 6,9 | 8,8 | 10,9 | 12,1 | 

|$X = \ln{x}$ | 0 | 2,77259 | 2,56495 | 3,82864 | 4,11087 | 4,33073 |
|:------------|:---:|:---:|:---:|:---:|:---:|:---:|
|$Y = \ln{y}$ |  -0,693147 | 1,38629 | 1,93152 | 2,17475 | 2,38876 | 2,49321 |

Таким образом, данная задача свелась к задаче 1.1.

Имеем,

$\sum\limits_{i=1}^{6} {\ln{x_i}} = 0 + 2,77259 + 2,56495 + 3,82864 + 4,11087 + 4,33073 = 17,60778$,

$\sum\limits_{i=1}^{6} {\ln{y_i}} =  -0,693147 + 1,38629 + 1,93152 + 2,17475 + 2,38876 + 2,49321 = 9,681383$,

$\sum\limits_{i=1}^{6} {\ln^2{x_i}} = 0^2 + 2,77259^2 + 2,56495^2 + 3,82864^2 + 4,11087^2 + 4,33073^2 = 64,57918255 $,

$\sum\limits_{i=1}^{6} {(\ln{x_i} \cdot \ln{y_i})}
 = 0\cdot (-0,693147) + 2,77259\cdot1,38629 + 2,56495\cdot 1,93152 + 3,82864\cdot 2,17475 + 4,11087\cdot 2,38876 + 4,33073\cdot 2,49321 = 37,7415 $.

Система уравнений будет иметь вид:

\begin{equation}
    \begin{cases}
        64,5792 a + 17,6078 b = 37,7415, \\
        17,6078 a + 6b = 9,6814.
    \end{cases}
\end{equation}

Определители системы:
\begin{equation}
    \Delta =
    \begin{vmatrix}
        64,5792  & 17,6078 \\
        17,6078 & 6
    \end{vmatrix} = 64,5792 \cdot 6 - 17,6078 \cdot 17,6078 = 77,4406,
\end{equation}
\begin{equation}
    \Delta_a =
    \begin{vmatrix}
        37,7415 & 17,607 \\
        9,6814 & 6
    \end{vmatrix} = 37,7415 \cdot 6 - 17,607 \cdot 9,6814 = 55,9886, 
    \quad
    \Delta_b =
    \begin{vmatrix}
        64,5792 &  37,7415 \\
        17,6078 &  9,6814 
    \end{vmatrix} = 64,5792 \cdot 9,6814 - 37,7415  \cdot 17,6078 = -39,3277. 
\end{equation}

Тогда,
\begin{equation}
    a = \cfrac{\Delta_a}{\Delta} = \cfrac{55,9886}{77,4406} = 0,723, 
    \quad
    b = \cfrac{\Delta_b}{\Delta} = \cfrac{-39,3277}{77,4406} = -0,508.
\end{equation}

Учитывая, что $b = \ln{\beta}$, находим $\beta = e^b = e^{-0,508} = 0,605$, 
и получаем искомую степенную функцию 
\begin{equation}
    y = 0,723 \cdot x^{0,605}.
\end{equation}

1.3  Найдем зависимость $y$ от $x$ в виде показательной функции $y = \beta \cdot e^{ax}$.


Прологарифмировав равенство $y = \beta \cdot e^{ax}$ по основанию $e$, получим
$\ln{y} = ax  + \ln{\beta}$.

Обозначим
$Y = \ln{y}$, $b = \ln{\beta}$.

Тогда получим линейную функцию $Y = a \cdot x + b$, где переменные $x$ и $Y$ связаны следующей табличной зависимостью:

|$x$ | 1 | 16 | 13 | 46 | 61 | 76 |
|:------------|:---:|:---:|:---:|:---:|:---:|:---:|
|$Y = \ln{y}$ |  -0,693147 | 1,38629 | 1,93152 | 2,17475 | 2,38876 | 2,49321 |

Таким образом, данная задача свелась к задаче 1.1.

Имеем,

$\sum\limits_{i=1}^{6} {x_i} = 1 + 16 + 13 + 46 + 61 + 76 = 213$,

$\sum\limits_{i=1}^{6} {x_i^2} = 1^2 + 16^2 + 13^2 + 46^2 + 61^2 + 76^2 = 1 + 256 + 169 + 2116 + 3721 + 5776 = 12039$,

$\sum\limits_{i=1}^{6} {\ln{y_i}} =  -0,693147 + 1,38629 + 1,93152 + 2,17475 + 2,38876 + 2,49321 = 9,681383$,

$\sum\limits_{i=1}^{6} {(x_i \cdot \ln{y_i})}
 = 1 \cdot -0,693147 + 16 \cdot 1,38629 + 13 \cdot 1,93152 + 46 \cdot 2,17475 + 61 \cdot 2,38876 + 76 \cdot 2,49321 = 481,834$.

Система уравнений будет иметь вид:

\begin{equation}
    \begin{cases}
        12039 a + 213 b = 481,834, \\
        213 a + 6b = 9,6814.
    \end{cases}
\end{equation}

Определители системы:
\begin{equation}
    \Delta =
    \begin{vmatrix}
        12039  & 213 \\
        213 & 6
    \end{vmatrix} = 12039 \cdot 6 - 213 \cdot 213 = 26865,
\end{equation}
\begin{equation}
    \Delta_a =
    \begin{vmatrix}
        481,834 & 213 \\
        9,6814 & 6
    \end{vmatrix} = 481,834 \cdot 6 - 213  \cdot 9,6814 = 828,866, 
    \quad
    \Delta_b =
    \begin{vmatrix}
        12039 &  481,834 \\
        213 &  9,6814 
    \end{vmatrix} = 12039 \cdot 9,6814 - 481,834 \cdot 213 = 13923,732.

Тогда,
\begin{equation}

Учитывая, что $b = \ln{\beta}$, находим $\beta = e^b = e^{0,5183} = 1,679$,

    y = 1,679 \cdot e^{0,0308x}.
\end{equation}
