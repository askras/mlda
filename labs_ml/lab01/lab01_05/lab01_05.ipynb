1.1  Найдем зависимость $y$ от $x$ в виде линейной функции $y = ax + b$.

Выберем значения коэффициентов $a$ и $b$ так, чтобы сумма квадратов отклонений
$ L(a, b) = \sum\limits_{i=1}^n (\Delta y_i)^2 = 
\sum\limits_{i=1}^{n} { \left((ax_i+b)-y_i\right)^2}$
была минимальной.

Имеем, 

$\sum\limits_{i=1}^{6} {x_i} = 1 + 3 + 5 + 7 + 9 + 11 = 36$,

$\sum\limits_{i=1}^{6} {y_i} =  2,0 + 10,1 + 22,6 + 37,1 + 54,5 + 73,2 = 199,5$,

$\sum\limits_{i=1}^{6} {x_i^2} = 1^2 + 3^2 + 5^2 + 7^2 + 9^2 + 11^2 = 286$,

$\sum\limits_{i=1}^{6} {(x_i \cdot y_i)} = 
1\cdot 2,0 + 3\cdot 10,1 + 5\cdot 22,6 + 7\cdot 37,1 + 9\cdot 54,5 + 11\cdot 73,2 = 1700,7$.

Тогда система уравнений примет вид:

\begin{equation}
    \begin{cases}
        286a + 36b = 1700,7, \\
        36a + 6b = 199,5.
    \end{cases}
\end{equation}

Решим систему уравнений по формулам Крамера:

\begin{equation}
    a = \cfrac{\Delta_a}{\Delta}, \;
    b = \cfrac{\Delta_b}{\Delta},
\end{equation}
где
\begin{equation}
    \Delta =
    \begin{vmatrix}
        286 & 36 \\
        36 & 6
    \end{vmatrix} = 286 \cdot 6 - 36 \cdot 36 = 420, 
\end{equation}
\begin{equation}
    \Delta_a =
    \begin{vmatrix}
        1700,7 & 36 \\
        199,5 & 6
    \end{vmatrix} = 1700,7 \cdot 6 - 36 \cdot 199,5 = 3022,2, 
    \quad
    \Delta_b =
    \begin{vmatrix}
        286 & 1700,7 \\
        36 & 199,5 
    \end{vmatrix} = 286 \cdot 199,5 - 1700,7 \cdot 36 = -4168,2. 
\end{equation}

Тогда,
\begin{equation}
    a = \cfrac{\Delta_a}{\Delta} = \cfrac{3022,2}{420} = 7,19, 
    \quad
    b = \cfrac{\Delta_b}{\Delta} = \cfrac{-4168,2}{420} = -9,92.
\end{equation}

Следовательно, искомая линейная функция будет иметь вид:
\begin{equation}
    y = 7,19 x - 9,92.
\end{equation}

1.2  Найдем зависимость $y$ от $x$ в виде степенной функции $y = \beta \cdot x^a$.

Прологарифмировав равенство $y = \beta \cdot x^a$ по основанию $e$, получим
$\ln{y} = a\cdot \ln{x} + \ln{\beta}$.

Обозначим
$Y = \ln{y}$, $X = \ln{x}$, $b = \ln{\beta}$.

Тогда получим линейную функцию $Y = a \cdot X + b$, где переменные $X$ и $Y$ связаны следующей табличной зависимостью:

|$X = \ln{x}$ | 0 | 1,0986 | 1,6094 | 1,9459 | 2,1972 | 2,3978 |
|:------------|:---:|:---:|:---:|:---:|:---:|:---:|
|$Y = \ln{y}$ |  0,6931 | 2,3125 | 3,1179 | 3,6136 | 3,9982 | 4,2931 |

Таким образом, данная задача свелась к задаче 1.1.

Имеем,

$\sum\limits_{i=1}^{6} {\ln{x_i}} = 0 + 1,0986 + 1,6094 + 1,9459 + 2,1972 + 2,3978 = 9,2489$,

$\sum\limits_{i=1}^{6} {\ln{y_i}} = 0,6931 + 2,3125 + 3,1179 + 3,6136 + 3,9982 + 4,2931 = 18,0284$,

$\sum\limits_{i=1}^{6} {\ln^2{x_i}} = 0^2 + 1,0986^2 + 1,6094^2 + 1,9459^2 + 2,1972^2 + 2,3978^2 = 18,1607$,

$\sum\limits_{i=1}^{6} {(\ln{x_i} \cdot \ln{y_i})}
 = 0\cdot 0,6931 + 1,0986\cdot 2,3125 + 3,1179\cdot 1,6094 + 1,9459\cdot 3,6136 + 2,1972\cdot 3,9982 + 2,3978\cdot 4,2931 = 33,6690$.

Система уравнений будет иметь вид:

\begin{equation}
    \begin{cases}
        18,1607 a + 9,2489 b = 33,6690, \\
        9,2489 a + 6b = 18,0284.
    \end{cases}
\end{equation}

Определители системы:
\begin{equation}
    \Delta =
    \begin{vmatrix}
        18,1607 & 9,2489 \\
        9,2489 & 6
    \end{vmatrix} = 18,1607 \cdot 6 - 9,2489 \cdot 9,2489 = 23,4220,
\end{equation}
\begin{equation}
    \Delta_a =
    \begin{vmatrix}
        33,6690 & 9,2489 \\
        18,0284 & 6
    \end{vmatrix} = 33,6690 \cdot 6 - 9,2489 \cdot 18,0284 = 35,2711, 
    \quad
    \Delta_b =
    \begin{vmatrix}
        18,1607 &  33,6690 \\
         9,2489 &  18,0284 
    \end{vmatrix} = 18,1607 \cdot 18,0284 - 33,6690 \cdot 9,2489 = 16,0071. 
\end{equation}

Тогда,
\begin{equation}
    a = \cfrac{\Delta_a}{\Delta} = \cfrac{35,2711}{23,4220} = 1,51, 
    \quad
    b = \cfrac{\Delta_b}{\Delta} = \cfrac{16,0071}{23,4220} = 0,68.
\end{equation}

Учитывая, что $b = \ln{\beta}$, находим $\beta = e^b = e^{0,68} = 1,97$, 
и получаем искомую степенную функцию 
\begin{equation}
    y = 1,97 \cdot x^{1,51}.
\end{equation}

1.3  Найдем зависимость $y$ от $x$ в виде показательной функции $y = \beta \cdot e^{ax}$.


Прологарифмировав равенство $y = \beta \cdot e^{ax}$ по основанию $e$, получим
$\ln{y} = ax  + \ln{\beta}$.

Обозначим
$Y = \ln{y}$, $b = \ln{\beta}$.

Тогда получим линейную функцию $Y = a \cdot x + b$, где переменные $x$ и $Y$ связаны следующей табличной зависимостью:
