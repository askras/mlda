## Задание

Дана таблица значений некоторой функциональной зависимости, полученной из $n = 6$ опытов

1. Методом наименьших квадратов по данной табличной зависимости найти аппроксимирующую функцию в виде:

  1.1) линейной функции $y = ax + b$;

  1.2) степенной функции $y = \beta \cdot x^a$;

  1.3) показательной функции $y = \beta \cdot e^{ax}$;

  1.4) квадратичной функции $y = ax^2 + bx + c$.

2. Построить в плоскости $xOy$ графики полученных функций и нанести экспериментальные точки.

3. Сравнить полученные результаты.


Промежуточные вычисления вести с точностью до 0,0001.

Значения параметров $a$, $b$, $c$ округлить до 0,01.

### Вариант индивидуального задания

Вариант 8

|$x$ | 2 | 5 | 8 | 11 | 14 | 17 |

1.1  Найдем зависимость $y$ от $x$ в виде линейной функции $y = ax + b$.

Выберем значения коэффициентов $a$ и $b$ так, чтобы сумма квадратов отклонений
$ L(a, b) = \sum\limits_{i=1}^n (\Delta y_i)^2 = 
\sum\limits_{i=1}^{n} { \left((ax_i+b)-y_i\right)^2}$
была минимальной.

Имеем, 

$\sum\limits_{i=1}^{6} {x_i} = 2 + 5 + 8 + 11 + 14 + 17 = 40$,

$\sum\limits_{i=1}^{6} {y_i} =  2,1 + 1,3 + 1,0 + 0,9 + 0,8 + 0,72 = 6,82$,

$\sum\limits_{i=1}^{6} {x_i^2} = 2^2 + 5^2 + 8^2 + 11^2 + 14^2 + 17^2 = 4 + 25 + 64 + 121 + 196 + 289 = 699$,

$\sum\limits_{i=1}^{6} {(x_i \cdot y_i)} = 
2\cdot 2,1 + 5\cdot 1,3 + 8\cdot 1,0 + 11\cdot 0,9 + 14\cdot 0,8 + 17\cdot 0,72 = 
4,2 + 6,5 + 8 + 9,9 + 11,2 + 12,24 = 52,04$.

Тогда система уравнений \eqref{eq:07} примет вид:

\begin{equation}
    \begin{cases}
        699a + 40b = 52,04, \\
        40a + 6b = 6,82.
    \end{cases}
\end{equation}

Решим систему уравнений по формулам Крамера:

\begin{equation}
    a = \cfrac{\Delta_a}{\Delta}, \;
    b = \cfrac{\Delta_b}{\Delta},
\end{equation}
где
\begin{equation}
    \Delta =
    \begin{vmatrix}
        699 & 40 \\
        40 & 6
    \end{vmatrix} = 699 \cdot 6 - 40 \cdot 40 = 2594, 
\end{equation}
\begin{equation}
    \Delta_a =
    \begin{vmatrix}
        52,04 & 40 \\
        6,82 & 6
    \end{vmatrix} = 52,04 \cdot 6 - 6,82 \cdot 40 = 39,44,
    \quad
    \Delta_b =
    \begin{vmatrix}
        699 & 52,04 \\
        40 & 6,82 
    \end{vmatrix} = 699 \cdot 6,82 - 40 \cdot 52,04 = 2685,58.
\end{equation}

Тогда,
\begin{equation}
    a = \cfrac{\Delta_a}{\Delta} = \cfrac{39,44}{2594} = 0,015, 
    \quad
    b = \cfrac{\Delta_b}{\Delta} = \cfrac{2685,58}{2594} = 1,035.
\end{equation}

Следовательно, искомая линейная функция будет иметь вид:
\begin{equation}
    y = 0,015x - 1,035.
\end{equation}

1.2  Найдем зависимость $y$ от $x$ в виде степенной функции $y = \beta \cdot x^a$.

Прологарифмировав равенство $y = \beta \cdot x^a$ по основанию $e$, получим
$\ln{y} = a\cdot \ln{x} + \ln{\beta}$.

Обозначим
$Y = \ln{y}$, $X = \ln{x}$, $b = \ln{\beta}$.

Тогда получим линейную функцию $Y = a \cdot X + b$, где переменные $X$ и $Y$ связаны следующей табличной зависимостью:

|$X = \ln{x}$ | 1 | 2,3219 | 3 | 3,4594 | 3,8074 | 4,088 |
|:------------|:---:|:---:|:---:|:---:|:---:|:---:|
|$Y = \ln{y}$ | 1,071 | 0,378 | 0 | -0,152 | -0,322 | -0,474 |

Таким образом, данная задача свелась к задаче 1.1.

Имеем,

$\sum\limits_{i=1}^{6} {\ln{x_i}} = 1 + 2,3219 + 3 + 3,4594 + 3,8074 + 4,088 = 17,6767$,

$\sum\limits_{i=1}^{6} {\ln{y_i}} = 1,071 + 0,378 + 0 + -0,152 + -0,322 + -0,474 = 0,501$,

$\sum\limits_{i=1}^{6} {\ln^2{x_i}} = 1^2 + 2,3219^2 + 3^2 + 3,4594^2 + 3,8074^2 + 4,088^2 = 58,5667 $,

$\sum\limits_{i=1}^{6} {(\ln{x_i} \cdot \ln{y_i})}
 = 1\cdot 1,071 + 2,3219\cdot 0,378 + 3 \cdot 0 + 3,4594 \cdot -0,152 + 3,8074 \cdot -0,322 + 4,088\cdot -0,474 = -1,7948$.

Система уравнений \eqref{eq:07} будет иметь вид:

\begin{equation}
    \begin{cases}
        58,5667 a + 17,6767 b = -1,7948, \\
        17,6767 a + 6b = 0,501.
    \end{cases}
\end{equation}

Определители системы:
\begin{equation}
    \Delta =
    \begin{vmatrix}
        58,5667  & 17,6767 \\
        17,6767 & 6
    \end{vmatrix} = 58,5667 \cdot 6 - 17,6767 \cdot 17,6767 = 38,9344,
\end{equation}
\begin{equation}
    \Delta_a =
    \begin{vmatrix}
        -1,7948 & 11,8139 \\
        0,501 & 6
    \end{vmatrix} = -1,7948 \cdot 6 - 11,8139  \cdot 0,501 = -16,6875,
    \quad
    \Delta_b =
    \begin{vmatrix}
        58,5667 &  -1,7948 \\
        17,6767 &  0,501 
    \end{vmatrix} = 58,5667 \cdot 0,501 - -1,7948  \cdot 17,6767 = 61,0608.
\end{equation}

Тогда,
\begin{equation}
    a = \cfrac{\Delta_a}{\Delta} = \cfrac{-16,6875}{38,9344} = -0,4286, 
    \quad
    b = \cfrac{\Delta_b}{\Delta} = \cfrac{61,0608}{38,9344} = 1,5683.
\end{equation}

Учитывая, что $b = \ln{\beta}$, находим $\beta = e^b = e^{1,5683} = 4,79848$, 
и получаем искомую степенную функцию 
\begin{equation}
    y = 4,79848 \cdot x^{-0,4286}.
\end{equation}

1.3  Найдем зависимость $y$ от $x$ в виде показательной функции $y = \beta \cdot e^{ax}$.


Прологарифмировав равенство $y = \beta \cdot e^{ax}$ по основанию $e$, получим
$\ln{y} = ax  + \ln{\beta}$.

Обозначим
$Y = \ln{y}$, $b = \ln{\beta}$.

Тогда получим линейную функцию $Y = a \cdot x + b$, где переменные $x$ и $Y$ связаны следующей табличной зависимостью:

|$x         $ | 2 | 5 | 8 | 11 | 14 | 17 |
|:------------|:---:|:---:|:---:|:---:|:---:|:---:|
|$Y = \ln{y}$ |  1,071 | 0,378 | 0 | -0,152 | -0,322 | -0,474 |

Таким образом, данная задача свелась к задаче 1.1.

Имеем,

$\sum\limits_{i=1}^{6} {x_i} = 2 + 5 + 8 + 11 + 14 + 17 = 40$,

$\sum\limits_{i=1}^{6} {x_i^2} = 2^2 + 5^2 + 8^2 + 11^2 + 14^2 + 17^2 = 4 + 25 + 64 + 121 + 196 + 289 = 699$,

$\sum\limits_{i=1}^{6} {\ln{y_i}} = 1,071 + 0,378 + 0 + -0,152 + -0,322 + -0,474 = 0,501$,

$\sum\limits_{i=1}^{6} {(x_i \cdot \ln{y_i})}
 = 2 \cdot 1,071 + 5 \cdot 0,378 + 8 \cdot 0 + 11 \cdot -0,152 + 14 \cdot -0,322 + 17 \cdot -0,474 = -10,206$.

Система уравнений \eqref{eq:07} будет иметь вид:

\begin{equation}
    \begin{cases}
        699 a + 40 b = -10,206, \\
        4 a + 6b = 0,501.
    \end{cases}
\end{equation}

Определители системы:
\begin{equation}
    \Delta =
    \begin{vmatrix}
        699  & 40 \\
        40 & 6
    \end{vmatrix} = 699 \cdot 6 - 40 \cdot 40 = 2594,
\end{equation}
\begin{equation}
    \Delta_a =
    \begin{vmatrix}
        -10,206 & 40 \\
        0,501 & 6
    \end{vmatrix} = -10,206 \cdot 6 - 40 \cdot 0,501 = -81,276, 
    \quad
    \Delta_b =
    \begin{vmatrix}
        699 & -10,206 \\
        40 & 0,501 
    \end{vmatrix} = 699 \cdot 0,501 - 40 \cdot -10,206 = 758,439. 
\end{equation}

Тогда,
\begin{equation}
    a = \cfrac{\Delta_a}{\Delta} = \cfrac{-81,276}{2593} = -0,031, 
    \quad
    b = \cfrac{\Delta_b}{\Delta} = \cfrac{758,439}{2593} = 0,2925.
\end{equation}

Учитывая, что $b = \ln{\beta}$, находим $\beta = e^b = e^{0,2925} = 1,33977$, 
и получаем искомую степенную функцию 
\begin{equation}
    y = 1,33977 \cdot e^{-0,031x}.
\end{equation}
