1.1  Найдем зависимость $y$ от $x$ в виде линейной функции $y = ax + b$.

Выберем значения коэффициентов $a$ и $b$ так, чтобы сумма квадратов отклонений
$ L(a, b) = \sum\limits_{i=1}^n (\Delta y_i)^2 = 
\sum\limits_{i=1}^{n} { \left((ax_i+b)-y_i\right)^2}$
была минимальной.

Имеем, 

$\sum\limits_{i=1}^{6} {x_i} = 25 + 40 + 55 + 70 + 85 + 100 = 375$,

$\sum\limits_{i=1}^{6} {y_i} =  2,4 + 3,2 + 3,8 + 4,3 + 4,7 + 5,1 = 23,5$,

$\sum\limits_{i=1}^{6} {x_i^2} = 25^2 + 40^2 + 55^2 + 70^2 + 85^2 + 100^2 = 625 + 1600 + 3025 + 4900 + 7225 + 10000 =27375$,

$\sum\limits_{i=1}^{6} {(x_i \cdot y_i)} = 
25\cdot 2,4+ 40\cdot 3,2 + 55\cdot 3,8 + 70\cdot 4,3 + 85\cdot 4,7 + 100\cdot 5,1 = 
60 + 128 + 209 + 301 + 399,5 + 510 = 1607,5$.

Тогда система уравнений \eqref{eq:07} примет вид:

\begin{equation}
    \begin{cases}
        27375a + 375b = 1607,5, \\
        375a + 6b = 23,5.
    \end{cases}
\end{equation}

Решим систему уравнений по формулам Крамера:

\begin{equation}
    a = \cfrac{\Delta_a}{\Delta}, \;
    b = \cfrac{\Delta_b}{\Delta},
\end{equation}
где
\begin{equation}
    \Delta =
    \begin{vmatrix}
        27375 & 375 \\
        375 & 6
    \end{vmatrix} = 27375 \cdot 6 - 375 \cdot 375 = 23625, 
\end{equation}
\begin{equation}
    \Delta_a =
    \begin{vmatrix}
        1607,5 & 375 \\
        23,5 & 6
    \end{vmatrix} = 1607,5 \cdot 6 - 375 \cdot 23,5 = 832,5, 
    \quad
    \Delta_b =
    \begin{vmatrix}
        27375 & 1607,5 \\
        375 & 23,5
    \end{vmatrix} = 27375 \cdot 23,5 - 1607,5 \cdot 375 = 40500. 
\end{equation}

Тогда,
\begin{equation}
    a = \cfrac{\Delta_a}{\Delta} = \cfrac{832,5}{23625} = 0,04, 
    \quad
    b = \cfrac{\Delta_b}{\Delta} = \cfrac{40500}{23625} = 1,71.
\end{equation}

Следовательно, искомая линейная функция будет иметь вид:
\begin{equation}
    y = 0,04 x + 1,71.
\end{equation}
