Промежуточные вычисления вести с точностью до 0,0001.

1.1  Найдем зависимость $y$ от $x$ в виде линейной функции $y = ax + b$.

Выберем значения коэффициентов $a$ и $b$ так, чтобы сумма квадратов отклонений
$ L(a, b) = \sum\limits_{i=1}^n (\Delta y_i)^2 = 
\sum\limits_{i=1}^{n} { \left((ax_i+b)-y_i\right)^2}$
была минимальной.

Имеем, 

$\sum\limits_{i=1}^{6} {x_i} = 10 + 14 + 18 + 22 + 26 + 30 = 120$,

$\sum\limits_{i=1}^{6} {y_i} =  4,2 + 4,5 + 4,8 + 5,1 + 5,2 + 5,4 = 29,2$,

$\sum\limits_{i=1}^{6} {x_i^2} = 10^2 + 14^2 + 18^2 + 22^2 + 26^2 + 30^2 = 2680$,

$\sum\limits_{i=1}^{6} {(x_i \cdot y_i)} = 
10\cdot 4,2 + 14\cdot 4,5 + 18\cdot 4,8 + 22\cdot 5,1 + 26\cdot 5,2 + 30\cdot 5,4 = 
600,8$.

Тогда система уравнений  примет вид:

\begin{equation}
    \begin{cases}
        2680a + 120b = 600,8, \\
        120a + 6b = 29,2.
    \end{cases}
\end{equation}

Решим систему уравнений по формулам Крамера:

\begin{equation}
    a = \cfrac{\Delta_a}{\Delta}, \;
    b = \cfrac{\Delta_b}{\Delta},
\end{equation}
где
\begin{equation}
    \Delta =
    \begin{vmatrix}
        2680 & 120 \\
        120 & 6
    \end{vmatrix} = 2680 \cdot 6 - 120 \cdot 120 = 1680, 
\end{equation}
\begin{equation}
    \Delta_a =
    \begin{vmatrix}
        600,8 & 120 \\
        29,2 & 6
    \end{vmatrix} = 100,8, 
    \quad
    \Delta_b =
    \begin{vmatrix}
        2680 & 600,8 \\
        120 & 29,2 
    \end{vmatrix} = 6160. 
\end{equation}

Тогда,
\begin{equation}
    a = \cfrac{\Delta_a}{\Delta} = \cfrac{100,8}{1680} = 0,06, 
    \quad
    b = \cfrac{\Delta_b}{\Delta} = \cfrac{6160}{1680} = 3,67.
\end{equation}

Следовательно, искомая линейная функция будет иметь вид:
\begin{equation}
    y = 0,06 x + 3,67.
\end{equation}

1.2  Найдем зависимость $y$ от $x$ в виде степенной функции $y = \beta \cdot x^a$.

Прологарифмировав равенство $y = \beta \cdot x^a$ по основанию $e$, получим
$\ln{y} = a\cdot \ln{x} + \ln{\beta}$.

Обозначим
$Y = \ln{y}$, $X = \ln{x}$, $b = \ln{\beta}$.

Тогда получим линейную функцию $Y = a \cdot X + b$, где переменные $X$ и $Y$ связаны следующей табличной зависимостью:

|$X = \ln{x}$ | 1 | 1,1461 | 1,2552 | 1,3424 | 1,4149 | 1,4771 |
|:------------|:---:|:---:|:---:|:---:|:---:|:---:|
|$Y = \ln{y}$ | 0,6232 | 0,6532 | 0,6812 | 0,7075 | 0,716 | 0,7323 |

Таким образом, данная задача свелась к задаче 1.1.

Имеем,

$\sum\limits_{i=1}^{6} {\ln{x_i}} =  7,6357$,

$\sum\limits_{i=1}^{6} {\ln{y_i}} = 4,1134$,

$\sum\limits_{i=1}^{6} {\ln^2{x_i}} = 9,1702$,

$\sum\limits_{i=1}^{6} {(\ln{x_i} \cdot \ln{y_i})}
 = 6,6138$.

Система уравнений будет иметь вид:

\begin{equation}
    \begin{cases}
        9,1702 a + 7,6357 b = 6,6138, \\
        7,6357 a + 6b = 4,1134.
    \end{cases}
\end{equation}

Определители системы:
\begin{equation}
    \Delta =
    \begin{vmatrix}
        9,1702  & 7,6357 \\
        7,6357 & 6
    \end{vmatrix} = -3,2827,
\end{equation}
\begin{equation}
    \Delta_a =
    \begin{vmatrix}
        6,6138 & 7,6357 \\
        4,1134 & 6
    \end{vmatrix} = 8,2741, 
    \quad
    \Delta_b =
    \begin{vmatrix}
        9,1702 &  6,6138 \\
        7,6357 &  4,1134 
    \end{vmatrix} = -12,7803. 
\end{equation}

Тогда,
\begin{equation}
    a = \cfrac{\Delta_a}{\Delta} = \cfrac{8,2741}{-3,2827} = -2,52, 
    \quad
    b = \cfrac{\Delta_b}{\Delta} = \cfrac{-12,7803}{-3,2827} = 3,89.
\end{equation}

Учитывая, что $b = \ln{\beta}$, находим $\beta = e^b = e^{3,89} = 48,91$, 
и получаем искомую степенную функцию 
\begin{equation}
    y = 48,91 \cdot x^{-2,52}.
\end{equation}

1.3  Найдем зависимость $y$ от $x$ в виде показательной функции $y = \beta \cdot e^{ax}$.


Прологарифмировав равенство $y = \beta \cdot e^{ax}$ по основанию $e$, получим
$\ln{y} = ax  + \ln{\beta}$.

Обозначим
$Y = \ln{y}$, $b = \ln{\beta}$.

Тогда получим линейную функцию $Y = a \cdot x + b$, где переменные $x$ и $Y$ связаны следующей табличной зависимостью:

|$x         $ | 10 | 14 | 18 | 22 | 26 | 30 |
|:------------|:---:|:---:|:---:|:---:|:---:|:---:|
|$Y = \ln{y}$ | 0,6232 | 0,6532 | 0,6812 | 0,7075 | 0,716 | 0,7323 |

Таким образом, данная задача свелась к задаче 1.1.

Имеем,

$\sum\limits_{i=1}^{6} {x_i} = 10 + 14 + 18 + 22 + 26 + 30 = 120$,

$\sum\limits_{i=1}^{6} {x_i^2} = 10^2 + 14^2 + 18^2 + 22^2 + 26^2 + 30^2 = 2680$,

$\sum\limits_{i=1}^{6} {\ln{y_i}} = 4,1134$,

$\sum\limits_{i=1}^{6} {(x_i \cdot \ln{y_i})} = 83,7884$.

Система уравнений будет иметь вид:

\begin{equation}
    \begin{cases}
        2680 a + 120 b = 83,7884, \\
        120 a + 6b = 4,1134.
    \end{cases}
\end{equation}

Тогда,
\begin{equation}
    a = 0,0054, 
    \quad
    b = 0,5769.
\end{equation}

Учитывая, что $b = \ln{\beta}$, находим $\beta = e^b = e^{0,5769} = 1,78051$, 
и получаем искомую показательную функцию 
\begin{equation}
    y = 1,78051 \cdot e^{0,0054x}.
\end{equation}

1.4 Найдем зависимость $y$ от $x$ в виде квадратичной функции $y = ax^2 + bx + c$.

Имеем,

$\sum\limits_{i=1}^{6} {x_i} = 10 + 14 + 18 + 22 + 26 + 30 = 120$,

$\sum\limits_{i=1}^{6} {x_i^2} = 10^2 + 14^2 + 18^2 + 22^2 + 26^2 + 30^2 = 2680$,

$\sum\limits_{i=1}^{6} {x_i^3} = 10^3 + 14^3 + 18^3 + 22^3 + 26^3 + 30^3 = 64800$,

$\sum\limits_{i=1}^{6} {x_i^4} = 10^4 + 14^4 + 18^4 + 22^4 + 26^4 + 30^4 = 1654624$,

$\sum\limits_{i=1}^{6} {y_i} =  4,2 + 4,5 + 4,8 + 5,1 + 5,2 + 5,4 = 29,2$,

$\sum\limits_{i=1}^{6} {(x_i \cdot y_i)} = 
10\cdot 4,2 + 14\cdot 4,5 + 18\cdot 4,8 + 22\cdot 5,1 + 26\cdot 5,2 + 30\cdot 5,4 = 600,8$.

$\sum\limits_{i=1}^{6} {(x_i^2 \cdot y_i)} = 13700,8$.

Система уравнений будет иметь вид:

\begin{equation}
    \begin{cases}
        1654624 a + 64800 b + 2680 c = 13700,8,\\
        64800 a + 2680 b + 120 c = 600,8,\\
        2680 a + 120 b + 6 c = 29,2.
    \end{cases}
\end{equation}

Решив систему уравнений получим,
\begin{equation}
    a = 0,001, 
    \quad
    b = 0,118,
    \quad
    с = 3,154.
\end{equation}

Следовательно, искомая квадратичная функция будет иметь вид:
\begin{equation}
    y = 0,001 x^2 + 0,118 x + 3,154.
\end{equation}

2. Построим в плоскости $xOy$ графики полученных функций и нанесем экспериментальные точки (Рисунок 2).

![](./img/lab1_10.png)

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-10, 10, 400)

y1 = 0.001 * x**2 - 0.118 * x + 3.154
y2 = 1.78051 * np.exp(0.0054 * x)
y3 = 0.06 * x + 3.67
y4 = 48.91 * x - 2.52

plt.figure(figsize=(10, 6))

plt.plot(x, y1, label='y = 0.001x^2 - 0.118x + 3.154', color='blue')
plt.plot(x, y2, label='y = 1.78051 * e^(0.0054x)', color='red')
plt.plot(x, y3, label='y = 0.06x + 3.67', color='green')
plt.plot(x, y4, label='y = 48.91x - 2.52', color='purple')

plt.xlim(-10, 10)
plt.ylim(-10, 10)

plt.xlabel('x')
plt.ylabel('y')

plt.title('Графики функций')

plt.legend()

plt.grid(True)

plt.show()
```
