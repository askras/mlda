```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

#DATASET: https://archive.ics.uci.edu/dataset/763/land+mines-1
```

```python
# 0-46 type 1 : null
# 47-93 type 2 : anti-tank
# 94-137 type 3 : antipersone
# 138-181 type 4 : booby-trapped
# 182-224 type 5 : M14 

df = pd.read_excel('MinesData.xlsx').to_numpy()
```

```python
print('Shape of the dataset:', df.shape)
print('Dataset type:', type(df))
print('A single row of the dataset is type of:', type(df[0]))
print('Types of elements:', type(df[0][0]), type(df[0][1]), type(df[0][2]), type(df[0][3]))
print('Dataset:')
print(df[0])
```

```python
V = [] # Voltage [0V..10,6V]
H = [] # High [0cm..20cm]
S = [] # Soil type [6 different types of soil]

for mine in df:
    V.append(mine[0])
    H.append(mine[1])
    S.append(mine[2])

    
```

```python
plt.figure(1)

n_1, = plt.plot(V[:46], H[:46], 'ro', label='1')
n_2, = plt.plot(V[47:94], H[47:94], 'go', label='2')
n_3, = plt.plot(V[94:138], H[94:138], 'bo', label='3')
n_4, = plt.plot(V[138:182], H[138:182], 'ko', label='4')
n_5, = plt.plot(V[182:225], H[182:225], 'yo', label='5')

#plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
plt.xlabel('V')
plt.ylabel('H')

plt.show()
```

```python
plt.figure(2)

n_1, = plt.plot(V[:46], S[:46], 'ro', label='1')
n_2, = plt.plot(V[47:94], S[47:94], 'go', label='2')
n_3, = plt.plot(V[94:138], S[94:138], 'bo', label='3')
n_4, = plt.plot(V[138:182], S[138:182], 'ko', label='4')
n_5, = plt.plot(V[182:225], S[182:225], 'yo', label='5')

#plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
plt.xlabel('V')
plt.ylabel('S')

plt.show()
```

```python
plt.figure(2)

n_1, = plt.plot(H[:46], S[:46], 'ro', label='1')
n_2, = plt.plot(H[47:94], S[47:94], 'go', label='2')
n_3, = plt.plot(H[94:138], S[94:138], 'bo', label='3')
n_4, = plt.plot(H[138:182], S[138:182], 'ko', label='4')
n_5, = plt.plot(H[182:225], S[182:225], 'yo', label='5')

#plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
plt.xlabel('H')
plt.ylabel('S')

plt.show()
```

## Conclusion
S - параметр с небольшим и более конкретным количеством возможных значений, поэтому графики получаются такими ровными
Зависимость параметров S и H дает мало информации из-за скудных вариаций данных параметров, точки налазят друг на друга. Можно выкинуть из анализа. Также противотанковые мины наиболее выражены своми признаками, их наиболее легко определить
