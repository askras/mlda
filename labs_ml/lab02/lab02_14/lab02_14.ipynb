## Использование химического анализа для определения происхождения вин

Мной был выбран набор данных под названием "Wine" по ссылке https://archive.ics.uci.edu/dataset/109/wine. Эти данные являются результатом химического анализа вин, выращенных в одном и том же регионе Италии, но полученных из трех разных сортов. В результате анализа определено 13 компонентов, обнаруженных в каждом из трех видов вин:

1. **Alcohol** - алкоголь
2. **Malicacid** - яблочная кислота
3. **Ash** - зола
4. **Alcalinity of ash** - щелочность золы
5. **Magnesium** - магний
6. **Total phenols** - общее количество фенолов
7. **Flavanoids** - флавоноиды
8. **Nonflavanoid phenols** - нефлавоноидные фенолы
9. **Proanthocyanins** - проантоцианы
10. **Color intensity** - интенсивность цвета
11. **Hue** - оттенок
12. **OD280OD315 of diluted wines** - концентрации разбавленных вин
13. **Proline** - пролин (аминокислота)

### Получение данных

```python
import numpy as np
data_path = "./wine/wine.data"
data = np.genfromtxt(data_path, delimiter=",")
print(data)
```

### Тип переменных и форма (shape)

```python
print ( "Data type : ", type(data) )
print ( "Data shape : ", data.shape )
print ( data[-4:] )
```

### Получение типа набора данных, строки, элемента

```python
data1 = np.genfromtxt(data_path, delimiter=",", dtype=None)
print('Shape of the dataset:', data1.shape)
print('Dataset type:', type(data1))
print('A single row of the dataset is type of:', type(data1[0]))
print('Types of elements:', type(data1[0][1]), type(data1[0][4]))
print('Dataset:')
print(data1)
```

### Указание типа столбцов при загрузке данных

```python
dt = np.dtype("f8, f8, f8, f8, f8, f8, f8, f8, f8, f8, f8, f8, f8, f8")
data2 = np.genfromtxt(data_path, delimiter=",", dtype=dt)
print('Shape of the dataset:', data2.shape)
print('Dataset type:', type(data2))
print('A single row of the dataset is type of:', type(data2[0]))
print('Types of elements:', type(data2[0][1]), type(data2[0][4]))
print('Dataset slice:')
print(data2[:10])
```

### Построение графиков с помощью matplotlib

```python
import matplotlib as mpl
import matplotlib.pyplot as plt
%matplotlib inline

# Data from individual columns
alcohol = [] # алкоголь
malicacid = [] # яблочная кислота
ash = [] # зола?
alcalinity_of_ash = [] # щелочность золы?
magnesium = [] # магний
total_phenols = [] # общее количество фенолов
flavanoids = [] # флавоноиды
nonflavanoid_phenols = [] # нефлавоноидные фенолы
proanthocyanins = [] # проантоцианы
color_intensity = [] # интенсивность цвета
hue = [] # оттенок
OD280_OD315_of_diluted_wines = [] # концентрации разбавленных вин
proline = [] # пролин - аминокислота

# Traversing the entire data2 collection
for dot in data2:
    alcohol.append(dot[1])
    malicacid.append(dot[2])
    ash.append(dot[3])
    alcalinity_of_ash.append(dot[4])
    magnesium.append(dot[5]) 
    total_phenols.append(dot[6])
    flavanoids.append(dot[7])
    nonflavanoid_phenols.append(dot[8])
    proanthocyanins.append(dot[9])
    color_intensity.append(dot[10])
    hue.append(dot[11])
    OD280_OD315_of_diluted_wines.append(dot[12])
    proline.append(dot[13]) 
plt.figure(1)

n_1, = plt.plot(malicacid[:50], proline[:50], 'ro', label='1')
n_2, = plt.plot(malicacid[50:100], proline[50:100], 'go', label='2')
n_3, = plt.plot(malicacid[100:150], proline[100:150], 'bo', label='3')
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
plt.xlabel('malicacid')
plt.ylabel('proline')

plt.show()
```

```python
plt.figure(2)

setosa, = plt.plot(alcohol[:50], hue[:50], 'ro', label='1')
versicolor, = plt.plot(alcohol[50:100], hue[50:100], 'go', label='2')
virginica, = plt.plot(alcohol[100:150], hue[100:150], 'bo', label='3')
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
plt.xlabel('alcohol')
plt.ylabel('hue')

plt.show()
```

```python
plt.figure(2)

setosa, = plt.plot(total_phenols[:50], malicacid[:50], 'ro', label='1')
versicolor, = plt.plot(total_phenols[50:100], malicacid[50:100], 'go', label='2')
virginica, = plt.plot(total_phenols[100:150], malicacid[100:150], 'bo', label='3')
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
plt.xlabel('total_phenols')
plt.ylabel('malicacid')

plt.show()
```

```python
plt.figure(2)

setosa, = plt.plot(alcohol[:50], flavanoids[:50], 'ro', label='1')
versicolor, = plt.plot(alcohol[50:100], flavanoids[50:100], 'go', label='2')
virginica, = plt.plot(alcohol[100:150], flavanoids[100:150], 'bo', label='3')
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
plt.xlabel('alcohol')
plt.ylabel('flavanoids')

plt.show()
```

### Выводы

Из графического распределения на построенных графиках видно, что первый сорт вина (изображен на графике красным цветом) хорошо отделяется, а сорта 2 и 3 представляют собой множества, разделение которых является непростой задачей. Данный набор данных размыто отображает различия сортов, поэтому для определения точного различия понадобится большее количество данных для сравнения (или же большее количество параметров для сравнения).
