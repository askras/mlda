##### 2.2. Форма набора данныx

```python
import pandas as pd
import numpy as np

DATASET_PATH = './datasets/world-data-2023.csv'

dtype = {
    'Country': str,
    'Density (P/Km2)': float,
    'Abbreviation': str,
    'Agricultural Land( %)': str,
    'Land Area(Km2)': float,
    'Armed Forces size': float,
    'Birth Rate': float,
    'Calling Code': float,
    'Capital/Major City': str,
    'Co2-Emissions': float,
    'CPI': float,
    'CPI Change (%)': str,
    'Currency-Code': str,
    'Fertility Rate': float,
    'Forested Area (%)': str,
    'Gasoline Price': str,
    'GDP': str,
    'Gross primary education enrollment (%)': str,
    'Gross tertiary education enrollment (%)': str,
    'Infant mortality': float,
    'Largest city': str,
    'Life expectancy': float,
    'Maternal mortality ratio': float,
    'Minimum wage': str,
    'Official language': str,
    'Out of pocket health expenditure': str,
    'Physicians per thousand': float,
    'Population': float,
    'Population: Labor force participation (%)': str,
    'Tax revenue (%)': str,
    'Total tax rate': str,
    'Unemployment rate': str,
    'Urban_population': float,
    'Latitude': float,
    'Longitude': float
}


df = pd.read_csv(DATASET_PATH, dtype=dtype, thousands=',')

dataset = df.to_numpy()

num_features = dataset.shape[1]

idxs = [i for i, v in enumerate(dataset[0]) if type(v) == float]
feats = [v for i, v in enumerate(dataset[0]) if type(v) == float]
                           
columns = dataset[:, idxs]



print("Number of features:", num_features, "\n")

i = 0
for feat in dtype:
    if dtype[feat] == float:  
        column = dataset[:, i]
        
        mean_values = np.nanmean(column, axis=0)
        max_values = np.nanmax(column, axis=0)
        min_values = np.nanmin(column, axis=0)
        is_empty = sum([v for v in pd.isnull(column)])
        
        print(f"----[{i}]{feat}----")
        print("Mean:", mean_values)
        print("Max:", max_values)
        print("Min:", min_values)
        print("Is empty:", is_empty)
    i+=1
```

На основе первичного анализа данных можно сделать вывод о том, что:
- В среднем на страну приходится 39 млн чел., из них 22 млн. прозивает в городской местности. (56 %)
- Средняя плотность населения на страну 356 чел./(км.^2)
- В среднем на старну приходится 689624 км.^2, а наибольшая занимаимая площадь составляет 17 млн.км^2 (Росссийская Федерация)
- Болшинство стран размещено в окресности с центром в 19.092351°, 20,232434° (Северная африка).

##### 2.3. Графические представления в разных плоскостях

```python
import matplotlib.pyplot as plt

def print_plot(feat_x: tuple, feat_y: tuple) -> None:
    x_label, x = feat_x
    y_label, y = feat_y
       
    plt.figure(figsize=(40, 40))
    plt.scatter(x, y)
    plt.xlabel(x_label)
    plt.ylabel(y_label)
    plt.show()

print_plot(
    ('Latitude', dataset[:, 33]),
    ('Population', dataset[:, 27]),
)
    
print_plot(
    ('Longitude', dataset[:, 34]),
    ('Population', dataset[:, 27]),
)

print_plot(
    ('Life expectancy', dataset[:, 21]),
    ('Population', dataset[:, 27]),
)

print_plot(
    ('Physicians per thousand', dataset[:, 26]),
    ('Maternal mortality ratio', dataset[:, 22]),
)

print_plot(
    ('Life expectancy', dataset[:, 21]),
    ('Urban_population', dataset[:, 32]),
)
```
