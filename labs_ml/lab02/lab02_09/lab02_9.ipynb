```python
import numpy as np
data_path = "./datasets/crx.data"
dt = np.dtype("U30, f8, f8, U30, U30, U30, U30, f8, U30, U30, i8, U30, U30, U30, i8, U30") 
data2 = np.genfromtxt(data_path, delimiter=",", dtype=dt)
print('Shape of the dataset:', data2.shape)
print('Dataset type:', type(data2))
print('A single row of the dataset is type of:', type(data2[0]))
print('Types of elements:', type(data2[0][1]), type(data2[0][4]))
print('Dataset slice:')
print(data2[:690])
```

```python
import matplotlib as mpl
import matplotlib.pyplot as plt
%matplotlib inline

# Data from individual columns
attribute1_length = [] 
attribute1_width = [] 
attribute2_length = [] 
attribute2_width = [] 

# Traversing the entire data2 collection
for dot in data2:
    attribute1_length.append(dot[1])
    attribute1_width.append(dot[2])
    attribute2_length.append(dot[1])
    attribute2_width.append(dot[2])

# We build graphs based on data projections
plt.figure(1)

approve, = plt.plot(attribute1_length[:70], attribute1_width[:70], 'g^', label='+')
deny, = plt.plot(attribute2_length[71:117], attribute2_width[71:117], 'ro', label='-')
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
plt.xlabel('Признак2')
plt.ylabel('Признак3')

plt.show()
```

По некоторому признаку видно, что количество одобрений и отказов на взятие кредита сосредоточено в большей степени на промежутке (20, 30), а чем больше Признак2, тем больше коэффициент отношения одобрений к отказам. Возможно признак "Признак2" обозначает зарплату заёмщика, чем выше зарплата, тем больше одобрений, в то же время со средней зарплатой людей берущих кредит больше, чем людей с высокой зарплатой.

```python
attribute3_length = [] 
attribute3_width = [] 
attribute4_length = [] 
attribute4_width = [] 

# Traversing the entire data2 collection
for dot in data2:
    attribute3_length.append(dot[10])
    attribute3_width.append(dot[5])
    attribute4_length.append(dot[10])
    attribute4_width.append(dot[5])

# We build graphs based on data projections
plt.figure(2)

approve, = plt.plot(attribute3_length[:70], attribute3_width[:70], 'g^', label='+')
deny, = plt.plot(attribute4_length[71:117], attribute4_width[71:117], 'ro', label='-')
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
plt.xlabel('Признак10')
plt.ylabel('Признак5')

plt.show()
```

Для Признака10 можно определить, что чем он больше, тем больше одобрений. По Признаку5 сложно что то сказать, поскольку соотношение одобрений к отказам близкое для многих значений Признака5
