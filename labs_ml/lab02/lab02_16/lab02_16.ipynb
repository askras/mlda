```python
import numpy as np
data_path = "./datasets/expanded"
data = np.genfromtxt(data_path, delimiter=",", dtype="U30")
print('Shape of the dataset:', data.shape)
print('Dataset type:', type(data))
print('A single row of the dataset is type of:', type(data[0]))
print('Types of elements:', type(data[0][1]))
print('Dataset slice:')
print(data[:10])
```

```python
import matplotlib as mpl
import matplotlib.pyplot as plt
%matplotlib inline

# Data from individual columns
cap_shape = [] #форма шляпки
cap_surface = [] # поверхность шляпки
cap_color = [] # цвет шляпки
bruises = [] # синяки?
odor = [] # запах
gill_attachment = [] # крепление жабр
gill_spacing = [] # расстояние между жабрами
gill_size = [] # размер жабер
gill_color = [] # цвет жабер 
stalk_shape = [] # форма стебля
stalk_root = [] # стеблевой корень
stalk_surface_above_ring = [] # поверхность стебля над кольцом
stalk_surface_below_ring = [] # поверхность стебля под кольцом
stalk_color_above_ring = [] # цвет стебля над кольцом
stalk_color_below_ring = [] # цвет стебля под кольцом
veil_type = [] # тип вуали
veil_color = [] # цвет вуали
ring_number = [] # число колец
ring_type = [] # тип колец
spore_print_color = [] # цвет спор
population = [] # популяция
habitat = [] # среда обитания


# Traversing the entire data collection
for dot in data:
    cap_shape.append(dot[1])
    cap_surface.append(dot[2])
    cap_color.append(dot[3])
    bruises.append(dot[4])
    odor.append(dot[5])
    gill_attachment.append(dot[6])
    gill_spacing.append(dot[7])
    gill_size.append(dot[8])
    gill_color.append(dot[9])
    stalk_shape.append(dot[10])
    stalk_root.append(dot[11])
    stalk_surface_above_ring.append(dot[12])
    stalk_surface_below_ring.append(dot[13])
    stalk_color_above_ring.append(dot[14])
    stalk_color_below_ring.append(dot[15])
    veil_type.append(dot[16])
    veil_color.append(dot[17])
    ring_number.append(dot[18])
    ring_type.append(dot[19])
    spore_print_color.append(dot[20])
    population.append(dot[21])
    habitat.append(dot[22])
```
