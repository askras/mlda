```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
data_path = "fatal-police-shootings-data.csv"
data = pd.read_csv(data_path)
data.head(10)
```

```python
data.info()
```

```python
data.columns
```

```python
data.hist();
```

```python
hist = data['age'].value_counts()
plt.bar(hist.index, hist);
```

```python
plt.bar(data.index, data['age'])
plt.show()
```

```python
sns.boxplot(data['age'])
```

```python
top_data = data[['state','age']]
top_data = top_data.groupby('state').sum()
top_data = top_data.sort_values('age', ascending=False)
top_data = top_data[:3].index.values
sns.boxplot(y='state', 
            x='age', 
            data=data[data.state.isin(top_data)], palette='Set2');
```

```python
hist = data['state'].value_counts()
plt.bar(hist.index, hist);
```

```python
data['gender'].value_counts()
```

```python
sns.countplot(x = 'gender', data = data);
```

```python
sns.countplot(x = 'state', data = data);
```

```python
sns.countplot(x = data[data['state'].isin(data['state'].value_counts().head(5).index)]['state'], data = data);
```

```python
feats=[f for f in data.columns if 'age' in f]
feats
```

```python
data[feats].hist(figsize=(5,5));
```

```python
sns.pairplot(data[feats]);
```

```python
sns.pairplot(data[feats+['gender']], hue='gender');
```
