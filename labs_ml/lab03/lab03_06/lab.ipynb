```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sns
```

```python
DATA_PATH = "./Billionaires Statistics Dataset.csv"
data = pd.read_csv(DATA_PATH)
data.head(10)
```

```python
data.info()
```

```python
data['age'].hist()
```

```python
sns.boxplot(data['age'])
```

```python
sns.countplot(x = data[data['country'].isin(data['country'].value_counts().head(5).index)]['country'], data = data);
```

```python
sns.countplot(x = data[data['category'].isin(data['category'].value_counts().head(3).index)]['category'], data = data);
```

```python
feats=[f for f in data.columns if 'enrollment' in f]
sns.pairplot(data[feats]);
```

```python
plt.scatter( 
    data['gross_primary_education_enrollment_country'],
    data['gross_tertiary_education_enrollment'],
    color='lightblue', edgecolors='blue'
)

plt.xlabel('gross_primary_education_enrollment_country')
plt.ylabel('gross_tertiary_education_enrollment')
```

```python
plt.scatter( 
    data['latitude_country'],
    data['longitude_country'],
    color='lightblue', edgecolors='blue'
)

plt.xlabel('latitude_country')
plt.ylabel('longitude_country')
```

```python
sns.heatmap(data.corr(numeric_only=True), cmap=plt.cm.PuBuGn);
```
