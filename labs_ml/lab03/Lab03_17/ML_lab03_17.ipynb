```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
data_path = "./Lab03/Electric_Vehicle_Population_Data.csv"
data = pd.read_csv(data_path)
data.head(10)
```

```python
data.info()
```

```python
data.columns
```

```python
data.hist();
#Создает гистограммы для всех числовых столбцов в наборе данных, что помогает визуализировать их распределение.
```

```python
hist = data['Legislative District'].value_counts()
#Считает количество уникальных значений в столбце 'Legislative District' и сохраняет результат в переменной hist.
plt.bar(hist.index, hist);
#Создает столбчатую диаграмму для данных, полученных из value_counts(),
#отображая количество каждого уникального значения в столбце 'Legislative District'.
```

```python
plt.bar(data.index, data['Legislative District'])
plt.show()
```

```python
sns.boxplot(data['Legislative District'])
#Создает ящичковую диаграмму (boxplot) для визуализации распределения данных в столбце 'Legislative District'
```

```python
import seaborn as sns

top_data = data[['City', 'Legislative District']]
#Создает новый набор данных, включающий столбцы 'City' и 'Legislative District'.
top_data = top_data.groupby('City').sum()
#Группирует данные по столбцу 'City' и суммирует значения 'Legislative District' для каждого города.
top_data = top_data.sort_values('Legislative District', ascending=False)
#Сортирует данные по убыванию значения 'Legislative District'.
top_data = top_data[:3].index.values
#Выбирает три города с наибольшими суммарными значениями 'Legislative District'.
sns.boxplot(y='City', x='Legislative District', 
            data=data[data.City.isin(top_data)], hue='City', palette='Set2', legend=False);
#Создает ящичковую диаграмму, где города из top_data выбраны как категории по оси y, а 'Legislative District' используется для оси x. 
#hue отвечает за разделение данных по городам с использованием цветов, а legend=False скрывает легенду.
```

```python
hist = data['City'].value_counts()
#Считает количество уникальных значений в столбце 'City'.
plt.bar(hist.index, hist);
#Создает столбчатую диаграмму, отображая количество записей для каждого уникального значения в столбце 'City'.
```

```python
data['Make'].value_counts()
#Считает количество уникальных значений в столбце 'Make'.
```

```python
sns.countplot(x = 'Make', data = data);
#Создает столбчатую диаграмму для визуализации количества записей для каждого уникального значения в столбце 'Make'.
```

```python
sns.countplot(x = 'City', data = data);
#Создает столбчатую диаграмму для визуализации количества записей для каждого уникального значения в столбце 'City'.
```

```python
sns.countplot(x = data[data['City'].isin(data['City'].value_counts().head(5).index)]['City'], data = data);
#Создает столбчатую диаграмму, отображая количество записей только для 5 наиболее часто встречающихся городов.
```

```python
feats=[f for f in data.columns if 'Legislative District' in f]
feats
#Создает список признаков, в названиях которых есть подстрока 'Legislative District'.
```

```python
data[feats].hist(figsize=(5,5));
#Создает гистограммы для всех признаков, перечисленных в списке feats.
```

```python
sns.pairplot(data[feats]);
# Создает графики рассеяния и гистограммы для всех пар признаков из списка feats.
```

```python
sns.pairplot(data[feats+['Make']], hue='Make');
#Создает графики рассеяния и гистограммы для всех пар признаков из списка feats, разделенных по цвету в соответствии с категорией 'Make'.
```
