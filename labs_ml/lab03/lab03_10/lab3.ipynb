```python
net_generation = initial_df.loc[initial_df['Features'] == 'net generation']

years = range(1980, 2022)


net_generation_fixed = net_generation.loc[(initial_df['2021'] != '--') & (initial_df['2021'] != 'NaN') & (initial_df['2021'] != 'ie')]
net_generation_fixed.loc[:, '2021'] = net_generation_fixed.loc[:, '2021'].astype(float)
# net_generation_fixed = net_generation
for year in years:
    net_generation_fixed[str(year)] = net_generation_fixed.loc[:, str(year)].replace('--', 'NaN')
    net_generation_fixed[str(year)] = net_generation_fixed.loc[:, str(year)].replace('ie', 'NaN')
    # net_generation_fixed = net_generation.loc[(initial_df[str(year)] != '--') & (initial_df[str(year)] != 'NaN') & (initial_df[str(year)] != 'ie')]
    net_generation_fixed.loc[:, str(year)] = net_generation_fixed.loc[:, str(year)].astype(float)



net_generation_countries = net_generation_fixed['Country'].unique().tolist()
```

```python
net_generation_fixed
```

```python
plot = net_generation_fixed.filter(['Country', '2021'], axis=1)
plot = plot.rename(columns={'2021': 'Net Generation'})
plot['Net Generation'] = plot['Net Generation'].astype(float)

plot = plot.sort_values(by='Net Generation', ascending=False).head(10)
plot_countries = plot['Country'].tolist()

plt.figure(figsize=(20, 10))
plt.bar(plot['Country'], plot['Net Generation'])
for country in plot_countries:
    country_df = plot.loc[plot['Country'] == country]
    plt.bar(country_df['Country'], country_df['Net Generation'], label=country)

plt.title('Net Generation of Electricity by Country in 2021')
plt.xlabel('Country')
plt.ylabel('Net Generation (GWh)')
plt.legend(loc='center left', bbox_to_anchor=(1, 0.5))
plt.show()
```

```python
plot = net_generation_fixed.sort_values(by='2021', ascending=False).head(10)
plot_countries = plot['Country'].tolist()

plt.figure(figsize=(20, 10))

for country in plot_countries:
    country_df = plot.loc[plot['Country'] == country]
    plt.plot(years, (country_df.iloc[0, 3:].astype(float)+2), label=country)
    
plt.title('Net Generation of Electricity by Country')
plt.xlabel('Year')
plt.ylabel('Net Generation (GWh) (Log Scale)')
plt.legend(loc='center left', bbox_to_anchor=(1, 0.5))
plt.show()
```

```python
net_consumption = initial_df.loc[initial_df['Features'] == 'net consumption']

net_consumption_fixed = net_consumption.loc[(initial_df['2021'] != '--') & (initial_df['2021'] != 'NaN') & (initial_df['2021'] != 'ie')]
net_consumption_fixed.loc[:, '2021'] = net_consumption_fixed.loc[:, '2021'].astype(float)
# net_consumption_fixed = net_generation
for year in years:
    net_consumption_fixed[str(year)] = net_consumption_fixed.loc[:, str(year)].replace('--', 'NaN')
    net_consumption_fixed[str(year)] = net_consumption_fixed.loc[:, str(year)].replace('ie', 'NaN')
    # net_consumption_fixed = net_generation.loc[(initial_df[str(year)] != '--') & (initial_df[str(year)] != 'NaN') & (initial_df[str(year)] != 'ie')]
    net_consumption_fixed.loc[:, str(year)] = net_consumption_fixed.loc[:, str(year)].astype(float)



net_consumption_countries = net_consumption_fixed['Country'].unique().tolist()



plot = net_consumption_fixed.sort_values(by='2021', ascending=False).head(5)
plot_countries = plot['Country'].tolist()

plt.figure(figsize=(20, 10))

for country in plot_countries:
    country_df = plot.loc[plot['Country'] == country]
    plt.plot(years, np.log(country_df.iloc[0, 3:].astype(float)+2), label=country)
    
plt.title('Net Consumption of Electricity by Country')
plt.xlabel('Year')
# plt.ylabel('Net Consumption (GWh) (Log Scale)')
plt.legend(loc='center left', bbox_to_anchor=(1, 0.5))
plt.show()
```
