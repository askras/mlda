## Задание

Дана таблица значений некоторой функциональной зависимости, полученной из $n = 6$ опытов

1. Методом наименьших квадратов по данной табличной зависимости найти аппроксимирующую функцию в виде:

  1.1) линейной функции $y = ax + b$;

  1.2) степенной функции $y = \beta \cdot x^a$;

  1.3) показательной функции $y = \beta \cdot e^{ax}$;

1.1  Найдем зависимость $y$ от $x$ в виде линейной функции $y = ax + b$.

Выберем значения коэффициентов $a$ и $b$ так, чтобы сумма квадратов отклонений
$ L(a, b) = \sum\limits_{i=1}^n (\Delta y_i)^2 = 
\sum\limits_{i=1}^{n} { \left((ax_i+b)-y_i\right)^2}$
была минимальной.

Имеем, 

$\sum\limits_{i=1}^{6} {x_i} = 3 + 5 + 7 + 9 + 11 + 13 = 48$,

$\sum\limits_{i=1}^{6} {y_i} =  26 + 76 + 150 + 240 + 360 + 500 = 1352$,

$\sum\limits_{i=1}^{6} {x_i^2} = 3^2 + 5^2 + 7^2 + 9^2 + 11^2 + 13^2 = 9 + 25 + 49 + 81 + 121 + 169 = 454$,

$\sum\limits_{i=1}^{6} {(x_i \cdot y_i)} = 
3\cdot 26 + 5\cdot 76 + 7\cdot 150 + 9\cdot 240 + 11\cdot 360 + 13\cdot 500 = 
78 + 380 + 1050 + 2160 + 3960 + 6500 = 14128$.

Тогда система уравнений \eqref{eq:07} примет вид:

\begin{equation}
    \begin{cases}
        454a + 48b = 14128, \\
        48a + 6b = 1352.
    \end{cases}
\end{equation}

Решим систему уравнений по формулам Крамера:

\begin{equation}
    a = \cfrac{\Delta_a}{\Delta}, \;
    b = \cfrac{\Delta_b}{\Delta},
\end{equation}
где
\begin{equation}
    \Delta =
    \begin{vmatrix}
        454 & 48 \\
        48 & 6
    \end{vmatrix} = 454 \cdot 6 - 48 \cdot 48 = 420, 
\end{equation}
\begin{equation}
    \Delta_a =
    \begin{vmatrix}
        14128 & 48 \\
        1352 & 6
    \end{vmatrix} = 14128 \cdot 6 - 1352 \cdot 48 = 19872,
    \quad
    \Delta_b =
    \begin{vmatrix}
        454 & 14128 \\
        48 & 1352 
    \end{vmatrix} = 454 \cdot 1352 - 48 \cdot 14128 = -64336.
\end{equation}

Тогда,
\begin{equation}
    a = \cfrac{\Delta_a}{\Delta} = \cfrac{19872}{420} = 47,31, 
    \quad
    b = \cfrac{\Delta_b}{\Delta} = \cfrac{-64336}{420} = -153,18.
\end{equation}

Следовательно, искомая линейная функция будет иметь вид:
\begin{equation}
    y = 47,31 x - 153,18.
\end{equation}

1.2  Найдем зависимость $y$ от $x$ в виде степенной функции $y = \beta \cdot x^a$.

Прологарифмировав равенство $y = \beta \cdot x^a$ по основанию $e$, получим
$\ln{y} = a\cdot \ln{x} + \ln{\beta}$.

Обозначим
$Y = \ln{y}$, $X = \ln{x}$, $b = \ln{\beta}$.

Тогда получим линейную функцию $Y = a \cdot X + b$, где переменные $X$ и $Y$ связаны следующей табличной зависимостью:

|$X = \ln{x}$ | 1,0986 | 1,6094 | 1,9459 | 2,1972 | 2,3979 | 2,5649 |
|:------------|:---:|:---:|:---:|:---:|:---:|:---:|
|$Y = \ln{y}$ | 3,2581 | 4,3307 | 5,0106 | 5,4806 | 5,8861 | 6,2146 |

Таким образом, данная задача свелась к задаче 1.1.

Имеем,

$\sum\limits_{i=1}^{6} {\ln{x_i}} = 1,0986 + 1,6094 + 1,9459 + 2,1972 + 2,3979 + 2,5649 = 11,8139$,

$\sum\limits_{i=1}^{6} {\ln{y_i}} = 3,2581 + 4,3307 + 5,0106 + 5,4806 + 5,8861 + 6,2146 = 30,1807$,

$\sum\limits_{i=1}^{6} {\ln^2{x_i}} = 1,0986^2 + 1,6094^2 + 1,9459^2 + 2,1972^2 + 2,3979^2 + 2,5649^2 = 
1,2069 + 2,5902 + 3,7865 + 4,8277 + 5,7499 + 6,5787 = 24,7399 $,

$\sum\limits_{i=1}^{6} {(\ln{x_i} \cdot \ln{y_i})}
 = 1,0986\cdot 3,2581 + 1,6094\cdot4,3307 + 1,9459\cdot 5,0106 + 2,1972\cdot 5,4806 + 2,3979\cdot 5,8861 + 2,5649\cdot 6,2146 =$
$ = 3,5793 + 6,9698 + 9,7501 + 12,04197 + 14,1143 + 15,9398 = 62,3953$.

Система уравнений \eqref{eq:07} будет иметь вид:

\begin{equation}
    \begin{cases}
        24,7399 a + 11,8139 b = 62,3953, \\
        11,8139 a + 6b = 30,1807.
    \end{cases}
\end{equation}

Определители системы:
\begin{equation}
    \Delta =
    \begin{vmatrix}
        24,7399  & 11,8139 \\
        11,8139 & 6
    \end{vmatrix} = 24,7399 \cdot 6 - 11,8139 \cdot 11,8139 = 8,8712,
\end{equation}
\begin{equation}
    \Delta_a =
    \begin{vmatrix}
        62,3953 & 11,8139 \\
        30,1807 & 6
    \end{vmatrix} = 62,3953 \cdot 6 - 11,8139  \cdot 30,1807 = 17,8198,
    \quad
    \Delta_b =
    \begin{vmatrix}
        24,7399 &  62,3953 \\
        11,8139 &  30,1807 
    \end{vmatrix} = 24,7399 \cdot 30,1807 - 11,8139  \cdot 62,39527 = 9,5360.
\end{equation}

Тогда,
\begin{equation}
    a = \cfrac{\Delta_a}{\Delta} = \cfrac{17,8198}{8,8712} = 2,01, 
    \quad
    b = \cfrac{\Delta_b}{\Delta} = \cfrac{9,5360}{8,8712} = 1,07.
\end{equation}

Учитывая, что $b = \ln{\beta}$, находим $\beta = e^b = e^{1,07} = 2,92$, 
и получаем искомую степенную функцию 
\begin{equation}
    y = 2,92 \cdot x^{2,01}.
\end{equation}

1.3  Найдем зависимость $y$ от $x$ в виде показательной функции $y = \beta \cdot e^{ax}$.
