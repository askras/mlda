## –ó–∞–¥–∞–Ω–∏–µ

–î–∞–Ω–∞ —Ç–∞–±–ª–∏—Ü–∞ –∑–Ω–∞—á–µ–Ω–∏–π –Ω–µ–∫–æ—Ç–æ—Ä–æ–π —Ñ—É–Ω–∫—Ü–∏–æ–Ω–∞–ª—å–Ω–æ–π –∑–∞–≤–∏—Å–∏–º–æ—Å—Ç–∏, –ø–æ–ª—É—á–µ–Ω–Ω–æ–π –∏–∑ $n = 6$ –æ–ø—ã—Ç–æ–≤

1. –ú–µ—Ç–æ–¥–æ–º –Ω–∞–∏–º–µ–Ω—å—à–∏—Ö –∫–≤–∞–¥—Ä–∞—Ç–æ–≤ –ø–æ –¥–∞–Ω–Ω–æ–π —Ç–∞–±–ª–∏—á–Ω–æ–π –∑–∞–≤–∏—Å–∏–º–æ—Å—Ç–∏ –Ω–∞–π—Ç–∏ –∞–ø–ø—Ä–æ–∫—Å–∏–º–∏—Ä—É—é—â—É—é —Ñ—É–Ω–∫—Ü–∏—é –≤ –≤–∏–¥–µ:

  1.1) –ª–∏–Ω–µ–π–Ω–æ–π —Ñ—É–Ω–∫—Ü–∏–∏ $y = ax + b$;

  1.2) —Å—Ç–µ–ø–µ–Ω–Ω–æ–π —Ñ—É–Ω–∫—Ü–∏–∏ $y = \beta \cdot x^a$;

  1.3) –ø–æ–∫–∞–∑–∞—Ç–µ–ª—å–Ω–æ–π —Ñ—É–Ω–∫—Ü–∏–∏ $y = \beta \cdot e^{ax}$;

  1.4) –∫–≤–∞–¥—Ä–∞—Ç–∏—á–Ω–æ–π —Ñ—É–Ω–∫—Ü–∏–∏ $y = ax^2 + bx + c$.

2. –ü–æ—Å—Ç—Ä–æ–∏—Ç—å –≤ –ø–ª–æ—Å–∫–æ—Å—Ç–∏ $xOy$ –≥—Ä–∞—Ñ–∏–∫–∏ –ø–æ–ª—É—á–µ–Ω–Ω—ã—Ö —Ñ—É–Ω–∫—Ü–∏–π –∏ –Ω–∞–Ω–µ—Å—Ç–∏ —ç–∫—Å–ø–µ—Ä–∏–º–µ–Ω—Ç–∞–ª—å–Ω—ã–µ —Ç–æ—á–∫–∏.

3. –°—Ä–∞–≤–Ω–∏—Ç—å –ø–æ–ª—É—á–µ–Ω–Ω—ã–µ —Ä–µ–∑—É–ª—å—Ç–∞—Ç—ã.


–ü—Ä–æ–º–µ–∂—É—Ç–æ—á–Ω—ã–µ –≤—ã—á–∏—Å–ª–µ–Ω–∏—è –≤–µ—Å—Ç–∏ —Å —Ç–æ—á–Ω–æ—Å—Ç—å—é –¥–æ 0,0001.

–ó–Ω–∞—á–µ–Ω–∏—è –ø–∞—Ä–∞–º–µ—Ç—Ä–æ–≤ $a$, $b$, $c$ –æ–∫—Ä—É–≥–ª–∏—Ç—å –¥–æ 0,01.

1.1  –ù–∞–π–¥–µ–º –∑–∞–≤–∏—Å–∏–º–æ—Å—Ç—å $y$ –æ—Ç $x$ –≤ –≤–∏–¥–µ –ª–∏–Ω–µ–π–Ω–æ–π —Ñ—É–Ω–∫—Ü–∏–∏ $y = ax + b$.

–í—ã–±–µ—Ä–µ–º –∑–Ω–∞—á–µ–Ω–∏—è –∫–æ—ç—Ñ—Ñ–∏—Ü–∏–µ–Ω—Ç–æ–≤ $a$ –∏ $b$ —Ç–∞–∫, —á—Ç–æ–±—ã —Å—É–º–º–∞ –∫–≤–∞–¥—Ä–∞—Ç–æ–≤ –æ—Ç–∫–ª–æ–Ω–µ–Ω–∏–π
$ L(a, b) = \sum\limits_{i=1}^n (\Delta y_i)^2 = 
\sum\limits_{i=1}^{n} { \left((ax_i+b)-y_i\right)^2}$
–±—ã–ª–∞ –º–∏–Ω–∏–º–∞–ª—å–Ω–æ–π.

–ò–º–µ–µ–º, 

$\sum\limits_{i=1}^{6} {x_i} = 10 + 14 + 18 + 22 + 26 + 30 = 120$,

$\sum\limits_{i=1}^{6} {y_i} =  4,2 + 4,5 + 4,8 + 5,1 + 5,2 + 5,4 = 29,2$,

$\sum\limits_{i=1}^{6} {x_i^2} = 10^2 + 14^2 + 18^2 + 22^2 + 26^2 + 30^2 = 2680$,

$\sum\limits_{i=1}^{6} {(x_i \cdot y_i)} = 
10\cdot 4,2 + 14\cdot 4,5 + 18\cdot 4,8 + 22\cdot 5,1 + 26\cdot 5,2 + 30\cdot 5,4 = 
600,8$.

–¢–æ–≥–¥–∞ —Å–∏—Å—Ç–µ–º–∞ —É—Ä–∞–≤–Ω–µ–Ω–∏–π  –ø—Ä–∏–º–µ—Ç –≤–∏–¥:

\begin{equation}
    \begin{cases}
        2680a + 120b = 600,8, \\
        120a + 6b = 29,2.
    \end{cases}
\end{equation}

–†–µ—à–∏–º —Å–∏—Å—Ç–µ–º—É —É—Ä–∞–≤–Ω–µ–Ω–∏–π –ø–æ —Ñ–æ—Ä–º—É–ª–∞–º –ö—Ä–∞–º–µ—Ä–∞:

\begin{equation}
    a = \cfrac{\Delta_a}{\Delta}, \;
    b = \cfrac{\Delta_b}{\Delta},
\end{equation}
–≥–¥–µ
\begin{equation}
    \Delta =
    \begin{vmatrix}
        2680 & 120 \\
        120 & 6
    \end{vmatrix} = 2680 \cdot 6 - 120 \cdot 120 = 1680, 
\end{equation}
\begin{equation}
    \Delta_a =
    \begin{vmatrix}
        600,8 & 120 \\
        29,2 & 6
    \end{vmatrix} = 100,8, 
    \quad
    \Delta_b =
    \begin{vmatrix}
        2680 & 600,8 \\
        120 & 29,2 
    \end{vmatrix} = 6160. 
\end{equation}

–¢–æ–≥–¥–∞,
\begin{equation}
    a = \cfrac{\Delta_a}{\Delta} = \cfrac{100,8}{1680} = 0,06, 
    \quad
    b = \cfrac{\Delta_b}{\Delta} = \cfrac{6160}{1680} = 3,67.
\end{equation}

–°–ª–µ–¥–æ–≤–∞—Ç–µ–ª—å–Ω–æ, –∏—Å–∫–æ–º–∞—è –ª–∏–Ω–µ–π–Ω–∞—è —Ñ—É–Ω–∫—Ü–∏—è –±—É–¥–µ—Ç –∏–º–µ—Ç—å –≤–∏–¥:
\begin{equation}
    y = 0,06 x + 3,67.
\end{equation}

1.2  –ù–∞–π–¥–µ–º –∑–∞–≤–∏—Å–∏–º–æ—Å—Ç—å $y$ –æ—Ç $x$ –≤ –≤–∏–¥–µ —Å—Ç–µ–ø–µ–Ω–Ω–æ–π —Ñ—É–Ω–∫—Ü–∏–∏ $y = \beta \cdot x^a$.

–ü—Ä–æ–ª–æ–≥–∞—Ä–∏—Ñ–º–∏—Ä–æ–≤–∞–≤ —Ä–∞–≤–µ–Ω—Å—Ç–≤–æ $y = \beta \cdot x^a$ –ø–æ –æ—Å–Ω–æ–≤–∞–Ω–∏—é $e$, –ø–æ–ª—É—á–∏–º
$\ln{y} = a\cdot \ln{x} + \ln{\beta}$.

–û–±–æ–∑–Ω–∞—á–∏–º
$Y = \ln{y}$, $X = \ln{x}$, $b = \ln{\beta}$.

–¢–æ–≥–¥–∞ –ø–æ–ª—É—á–∏–º –ª–∏–Ω–µ–π–Ω—É—é —Ñ—É–Ω–∫—Ü–∏—é $Y = a \cdot X + b$, –≥–¥–µ –ø–µ—Ä–µ–º–µ–Ω–Ω—ã–µ $X$ –∏ $Y$ —Å–≤—è–∑–∞–Ω—ã —Å–ª–µ–¥—É—é—â–µ–π —Ç–∞–±–ª–∏—á–Ω–æ–π –∑–∞–≤–∏—Å–∏–º–æ—Å—Ç—å—é:

|$Y = \ln{y}$ | 0,6232 | 0,6532 | 0,6812 | 0,7075 | 0,716 | 0,7323 |

–¢–∞–∫–∏–º –æ–±—Ä–∞–∑–æ–º, –¥–∞–Ω–Ω–∞—è –∑–∞–¥–∞—á–∞ —Å–≤–µ–ª–∞—Å—å –∫ –∑–∞–¥–∞—á–µ 1.1.

–ò–º–µ–µ–º,

$\sum\limits_{i=1}^{6} {\ln{x_i}} =  7,6357$,

$\sum\limits_{i=1}^{6} {\ln{y_i}} = 4,1134$,

$\sum\limits_{i=1}^{6} {\ln^2{x_i}} = 9,1702$,

$\sum\limits_{i=1}^{6} {(\ln{x_i} \cdot \ln{y_i})}
 = 6,6138$.

–°–∏—Å—Ç–µ–º–∞ —É—Ä–∞–≤–Ω–µ–Ω–∏–π –±—É–¥–µ—Ç –∏–º–µ—Ç—å –≤–∏–¥:

\begin{equation}
    \begin{cases}
        9,1702 a + 7,6357 b = 6,6138, \\
        7,6357 a + 6b = 4,1134.
    \end{cases}
\end{equation}

–û–ø—Ä–µ–¥–µ–ª–∏—Ç–µ–ª–∏ —Å–∏—Å—Ç–µ–º—ã:
\begin{equation}
    \Delta =
    \begin{vmatrix}
        9,1702  & 7,6357 \\
        7,6357 & 6
    \end{vmatrix} = -3,2827,
\end{equation}
\begin{equation}
    \Delta_a =
    \begin{vmatrix}
        6,6138 & 7,6357 \\

    \end{vmatrix} = 8,2741, 
    \quad
    \Delta_b =
    \begin{vmatrix}
        9,1702 &  6,6138 \\
        7,6357 &  4,1134 
    \end{vmatrix} = -12,7803. 
\end{equation}

–¢–æ–≥–¥–∞,
\begin{equation}
    a = \cfrac{\Delta_a}{\Delta} = \cfrac{8,2741}{-3,2827} = -2,52, 
    \quad
    b = \cfrac{\Delta_b}{\Delta} = \cfrac{-12,7803}{-3,2827} = 3,89.
\end{equation}

–£—á–∏—Ç—ã–≤–∞—è, —á—Ç–æ $b = \ln{\beta}$, –Ω–∞—Ö–æ–¥–∏–º $\beta = e^b = e^{3,89} = 48,91$, 
–∏ –ø–æ–ª—É—á–∞–µ–º –∏—Å–∫–æ–º—É—é —Å—Ç–µ–ø–µ–Ω–Ω—É—é —Ñ—É–Ω–∫—Ü–∏—é 
\begin{equation}

1.3  –ù–∞–π–¥–µ–º –∑–∞–≤–∏—Å–∏–º–æ—Å—Ç—å $y$ –æ—Ç $x$ –≤ –≤–∏–¥–µ –ø–æ–∫–∞–∑–∞—Ç–µ–ª—å–Ω–æ–π —Ñ—É–Ω–∫—Ü–∏–∏ $y = \beta \cdot e^{ax}$.


–ü—Ä–æ–ª–æ–≥–∞—Ä–∏—Ñ–º–∏—Ä–æ–≤–∞–≤ —Ä–∞–≤–µ–Ω—Å—Ç–≤–æ $y = \beta \cdot e^{ax}$ –ø–æ –æ—Å–Ω–æ–≤–∞–Ω–∏—é $e$, –ø–æ–ª—É—á–∏–º

–¢–∞–∫–∏–º –æ–±—Ä–∞–∑–æ–º, –¥–∞–Ω–Ω–∞—è –∑–∞–¥–∞—á–∞ —Å–≤–µ–ª–∞—Å—å –∫ –∑–∞–¥–∞—á–µ 1.1.

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-10, 10, 400)

y1 = 0.001 * x**2 - 0.118 * x + 3.154
y2 = 1.78051 * np.exp(0.0054 * x)
y3 = 0.06 * x + 3.67
y4 = 48.91 * x - 2.52

plt.figure(figsize=(10, 6))

plt.plot(x, y1, label='y = 0.001x^2 - 0.118x + 3.154', color='blue')
plt.plot(x, y2, label='y = 1.78051 * e^(0.0054x)', color='red')
plt.plot(x, y3, label='y = 0.06x + 3.67', color='green')
plt.plot(x, y4, label='y = 48.91x - 2.52', color='purple')

plt.xlim(-10, 10)
plt.ylim(-10, 10)

plt.xlabel('x')
plt.ylabel('y')

plt.title('–ì—Ä–∞—Ñ–∏–∫–∏ —Ñ—É–Ω–∫—Ü–∏–π')

plt.legend()

plt.grid(True)

plt.show()
```
