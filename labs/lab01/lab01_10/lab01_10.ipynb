# Линейная функция

1.1  Найдем зависимость $y$ от $x$ в виде линейной функции $y = ax + b$.

Выберем значения коэффициентов $a$ и $b$ так, чтобы сумма квадратов отклонений
$ L(a, b) = \sum\limits_{i=1}^n (\Delta y_i)^2 = 
\sum\limits_{i=1}^{n} { \left((ax_i+b)-y_i\right)^2}$
была минимальной.

Имеем, 

$\sum\limits_{i=1}^{6} {x_i} = 3 + 4 + 5 + 6 + 7 + 8 = 33$,

$\sum\limits_{i=1}^{6} {y_i} =  13 + 31 + 64 + 105 + 170 + 252 = 635$,

$\sum\limits_{i=1}^{6} {x_i^2} = 199$,

$\sum\limits_{i=1}^{6} {(x_i \cdot y_i)} = 4319$.

Тогда система уравнений примет вид:

\begin{equation}
    \begin{cases}
        199a + 33b = 4319 \\
        33a + 6b = 635.
    \end{cases}
\end{equation}

Решим систему уравнений по формулам Крамера:

\begin{equation}
    a = \cfrac{\Delta_a}{\Delta}, \;
    b = \cfrac{\Delta_b}{\Delta},
\end{equation}
где
\begin{equation}
    \Delta =
    \begin{vmatrix}
        199 & 33 \\
        33 & 6
    \end{vmatrix} = 199 \cdot 6 - 33\cdot 33 = 105, 
\end{equation}
\begin{equation}
    \Delta_a =
    \begin{vmatrix}
        4319 & 33 \\
        635 & 6
    \end{vmatrix} = 4319 \cdot 6 - 33 \cdot 635 = 4959, 
    \quad
    \Delta_b =
    \begin{vmatrix}
        199 & 4319 \\
        33 & 635 
    \end{vmatrix} = 199 \cdot 635 - 4319 \cdot 33 = 16162. 
\end{equation}

Тогда,
\begin{equation}
    a = \cfrac{\Delta_a}{\Delta} = \cfrac{4959}{105} = 47,2286, 
    \quad
    b = \cfrac{\Delta_b}{\Delta} = \cfrac{16162}{105} = -153,924.
\end{equation}

Следовательно, искомая линейная функция будет иметь вид:
\begin{equation}
    y = 47,2286 x - 153,924.
\end{equation}

# Степенная функция

1.2  Найдем зависимость $y$ от $x$ в виде степенной функции $y = \beta \cdot x^a$.

Прологарифмировав равенство $y = \beta \cdot x^a$ по основанию $e$, получим
$\ln{y} = a\cdot \ln{x} + \ln{\beta}$.

Обозначим
$Y = \ln{y}$, $X = \ln{x}$, $b = \ln{\beta}$.

Тогда получим линейную функцию $Y = a \cdot X + b$, где переменные $X$ и $Y$ связаны следующей табличной зависимостью:

|$X = \ln{x}$ | 1,0986 | 1,3863 | 1,6094 | 1,7918 | 1,9459 | 2,0794 |
|:------------|:---:|:---:|:---:|:---:|:---:|:---:|
|$Y = \ln{y}$ |  2,5649 | 3,4339 | 4,1589 | 4,6539 | 5,1358 | 5,5294 |

Таким образом, данная задача свелась к задаче 1.1.

Имеем,

$\sum\limits_{i=1}^{6} {\ln{x_i}} = 9,9114$,

$\sum\limits_{i=1}^{6} {\ln{y_i}} = 25,4768$,

$\sum\limits_{i=1}^{6} {\ln^2{x_i}} = 17,0399$,

$\sum\limits_{i=1}^{6} {(\ln{x_i} \cdot \ln{y_i})}
 = 44,1020$.

Система уравнений будет иметь вид:

\begin{equation}
    \begin{cases}
        17,0399 a + 9,9114 b = 44,1020, \\
        9,9114 a + 6b = 25,4768.
    \end{cases}
\end{equation}

Определители системы:
\begin{equation}
    \Delta =
    \begin{vmatrix}
        17,0399  & 9,9114 \\
        9,9114 & 6
    \end{vmatrix} = 17,0399 \cdot 6 - 9,9114 \cdot 9,9114 = 4,0036,
\end{equation}
\begin{equation}
    \Delta_a =
    \begin{vmatrix}
        44,1020 & 6,5792 \\
        25,4768 & 6
    \end{vmatrix} = 12,1012, 
    \quad
    \Delta_b =
    \begin{vmatrix}
        17,0399 &  44,1020 \\
        9,9114 &  25,4768 
    \end{vmatrix} = -2,9903. 
\end{equation}

Тогда,
\begin{equation}
    a = \cfrac{\Delta_a}{\Delta} = \cfrac{12,1012}{4,0036} = 3,0026, 
    \quad
    b = \cfrac{\Delta_b}{\Delta} = \cfrac{-2,9903}{4,0036} = -0,7469.
\end{equation}

Учитывая, что $b = \ln{\beta}$, находим $\beta = e^b = e^{-0,7469} = 0,4738$, 
и получаем искомую степенную функцию 
\begin{equation}
    y = 0,4738 \cdot x^{3,0026}.
\end{equation}
