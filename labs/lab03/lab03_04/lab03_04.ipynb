```python
import numpy as np 
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sns
%matplotlib inline

data_path = "atp_matches_qual_chall_2023.csv"
data = pd.read_csv(data_path)
pd.set_option('display.max_columns', None)  
data.head(20)  
```

```python
data.info()
```

```python
data.columns
```

```python
selected_columns = ['winner_age', 'winner_seed', 'winner_ht', 'winner_rank_points']

# Постройте гистограммы для выбранных столбцов
data[selected_columns].hist(bins=10, figsize=(15, 10))
plt.suptitle("", fontsize=16)
plt.show()
```

```python
hist = data['winner_age'].value_counts()
plt.bar(hist.index, hist);
```

```python
plt.bar(data.index, data['winner_age'])
plt.show()
```

```python
sns.boxplot(data['winner_age'])
```

```python
top_data = data[['winner_ioc','winner_age']]
top_data = top_data.groupby('winner_ioc').sum()
top_data = top_data.sort_values('winner_age', ascending=False)
top_data = top_data[:5].index.values
sns.boxplot(y='winner_ioc', 
            x='winner_age', 
            data=data[data.winner_ioc.isin(top_data)], palette='Set2');
```

```python
import matplotlib.pyplot as plt
top_countries = data['winner_ioc'].value_counts().head(10)
plt.figure(figsize=(12, 6))
plt.bar(top_countries.index, top_countries)
plt.title("Гистограмма победителей по странам (Топ 10)")
plt.xlabel("Страна")
plt.ylabel("Количество победителей")
plt.xticks(rotation=45)
plt.show()
```

```python
import matplotlib.pyplot as plt

# Фильтрация данных для победителей из России (код страны "RUS")
russian_winners = data[data['winner_ioc'] == 'RUS']

# Разделите столбец winner_name на фамилию и имя
russian_winners[['last_name', 'first_name']] = russian_winners['winner_name'].str.split(' ', n=1, expand=True)

# Создайте гистограмму на основе фамилий и первых букв имен победителей из России
plt.figure(figsize=(12, 6))
plt.title("Гистограмма победителей из России")
plt.xlabel("Фамилия и имя")
plt.ylabel("Количество")

# Объедините фамилии и имена
name_parts_combined = russian_winners['last_name'] + ' ' + russian_winners['first_name']

# Постройте гистограмму
name_parts_combined.value_counts().plot(kind='bar')
plt.xticks(rotation=45)  # Поверните метки оси x для лучшей читаемости

plt.show()
```
