```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset (replace with the actual path to the downloaded CSV file)
data_path = "./Hopsital Dataset.csv"
data = pd.read_csv(data_path)
data.head(10)
```

```python
# Bar plot for gender distribution
plt.figure(figsize=(8, 6))
sns.countplot(x='Gender', data=data)
plt.title('Gender Distribution')
plt.xlabel('Gender')
plt.ylabel('Count')
plt.show()

# Bar plot for diagnosis
plt.figure(figsize=(12, 6))
sns.countplot(x='Name of Drug', data=data, order=data['Name of Drug'].value_counts().index)
plt.title('Diagnosis Distribution')
plt.xlabel('Name of Drug')
plt.ylabel('Count')
plt.xticks(rotation=90)
plt.show()

# Histogram for age
plt.figure(figsize=(30, 6))
plt.hist(data['Age'], bins=10, edgecolor='black')
plt.title('Age Distribution')
plt.xlabel('Age')
plt.ylabel('Count')
plt.show()

top_data = data[['Age','Gender']]
top_data = top_data.groupby('Age').sum()
top_data = top_data.sort_values('Gender', ascending=False)
top_data = top_data[:10].index.values
sns.boxplot(y='Age', 
            x='Gender', 
            data=data[data.Age.isin(top_data)], palette='Set2');
```
