### Найдем лучших актёров

```python
actors_ratings = data.groupby(['Stars'])['Movie Rating'].mean()
sorted_actors_ratings = actors_ratings.sort_values(ascending=False)
top_actors = sorted_actors_ratings.head(10)
top_actors
```

### Найдём лучших режиссёров

```python
directors_ratings = data.groupby(['Director'])['Movie Rating'].mean()
sorted_directors_ratings = directors_ratings.sort_values(ascending=False)
top_directors = sorted_directors_ratings.head(10)
top_directors
```

### Зависимоть рейтинга от длины фильма

```python
plt.figure(figsize=(10, 6))
plt.scatter(data['Run Time in minutes'], data['Movie Rating'], alpha=0.5)
plt.title('Зависимость рейтинга фильма от длительности')
plt.xlabel('Длительность фильма (минуты)')
plt.ylabel('Рейтинг фильма')
plt.grid(True)

# Отобразите график
plt.show()
```

### Зависимость средней оценки от года выпуска фильма

```python
average_ratings_by_year = data.groupby('Year of Release')['Movie Rating'].mean()
plt.figure(figsize=(12, 6))
plt.plot(average_ratings_by_year.index, average_ratings_by_year.values, marker='o', linestyle='-')
plt.title('Зависимость средней оценки от года выпуска')
plt.xlabel('Год выпуска')
plt.ylabel('Средняя оценка')
plt.grid(True)
plt.show()
```

### Средняя длительность фильмов по годам выпуска

```python
average_duration_by_year = data.groupby('Year of Release')['Run Time in minutes'].mean()
plt.figure(figsize=(12, 6))
plt.plot(average_duration_by_year.index, average_duration_by_year.values, marker='o', linestyle='-')
plt.title('Средняя длительность фильмов по годам выпуска')
plt.xlabel('Год выпуска')
plt.ylabel('Средняя длительность (минуты)')
plt.grid(True)
plt.show()
```

### Количество фильмов каждого жанра

```python
data['Genre'] = data['Genre'].str.split(', ')
all_genres = [genre for sublist in data['Genre'] for genre in sublist]
genre_counts = pd.Series(all_genres).value_counts()
plt.figure(figsize=(12, 6))
genre_counts.plot(kind='bar')
plt.title('Количество фильмов каждого жанра')
plt.xlabel('Жанр')
plt.ylabel('Количество фильмов')
plt.grid(axis='y')
plt.xticks(rotation=60)
plt.show()
```

### Матрица корреляции числовых атрибутов

```python
numeric_attributes = data[['Movie Rating', 'Run Time in minutes', 'Votes', 'MetaScore', 'Gross']]
correlation_matrix = numeric_attributes.corr()
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', linewidths=0.5)
plt.title('Матрица корреляции числовых атрибутов')
plt.show()
```

```python
plt.figure(figsize=(8, 8))
plt.pie(genre_counts, labels=genre_counts.index, autopct='%1.1f%%', startangle=140)
plt.title('Доля каждого жанра среди всех фильмов')
plt.legend(loc='best')
plt.show()
```

### Зависимость рейтинга от длины имени фильма)

```python
data['Name_Length'] = data['Movie Name'].apply(len)
plt.figure(figsize=(10, 6))
plt.scatter(data['Name_Length'], data['Movie Rating'], alpha=0.5)
plt.title('Зависимость рейтинга от длины имени фильма')
plt.xlabel('Длина имени фильма')
plt.ylabel('Рейтинг фильма')
plt.grid(True)
plt.show()
```
