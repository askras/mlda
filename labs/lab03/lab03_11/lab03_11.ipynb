Этот датасет содержит информацию о популярных блогерах на YouTube, и он был получен с использованием платформы HypeAuditor, которая занимается анализом и статистикой для стримеров и создателей контента на различных социальных платформах. Датасет предоставляет информацию о различных критериях, связанных с производительностью и присутствием стримеров на YouTube. Вот описание этих критериев:

1. Ранг (Rank): Этот критерий указывает на позицию или рейтинг стримера среди топовых стримеров на YouTube. Это важно для определения их относительной популярности и влияния.

2. Имя пользователя (Username): Имя пользователя на YouTube позволяет уникально идентифицировать каждого стримера, таким образом, обеспечивая их идентификацию и поиск на платформе.

3. Категории (Categories): Этот критерий предоставляет информацию о категориях контента, в которых стример размещает свои видеоролики. Это помогает понять, о чем идет речь в их видео и к какой аудитории они обращаются.

4. Подписчики (Subscribers): Этот параметр указывает на среднее количество подписчиков на канале стримера. Это важный показатель, отражающий размер и преданность их аудитории.

5. Страна (Country): Указывает на страну, где находится стример. Это может быть полезным для понимания аудитории и местоположения создателя контента.

6. Посещения (Visits): Этот параметр записывает среднее количество просмотров на канале стримера. Это помогает определить, насколько часто просматривают их видеоролики.

7. Лайки (Likes): Этот критерий указывает на среднее количество "Лайков" (понравившихся) на видеороликах стримера. Он отражает уровень увлечения и одобрения аудитории.

8. Комментарии (Comments): Этот параметр показывает среднее количество комментариев, оставленных на видеороликах стримера. Это важно для оценки вовлеченности и взаимодействия с аудиторией.

9. Ссылки (Links): Предоставляет ссылки на YouTube-каналы стримеров, обеспечивая прямой доступ к их контенту.

Этот датасет может быть полезен в нескольких областях:

1. Анализ трендов: Он может использоваться для выявления новых трендов в самых популярных категориях контента и для отслеживания роста новых блогеров на платформе.

2. Изучение аудитории: Датасет помогает понять, где находится аудитория стримеров и какие предпочтения в контенте существуют в разных регионах.

3. Маркетинговая стратегия: Бренды и компании могут использовать данные для выбора подходящих стримеров для сотрудничества и маркетинговых кампаний на основе их производительности.

4. Бенчмаркинг: Стримеры могут сравнивать свою производительность с другими создателями для определения своей конкурентоспособности.

5. Исследование сообщества создателей контента: Этот датасет может использоваться для изучения сообщества создателей контента YouTube и его влияния на платформу.

6. Рекомендации по контенту: Платформы рекомендаций могут использовать данные для улучшения рекомендаций видеороликов на основе категорий и производительности.

7. Анализ вовлеченности аудитории: Данные о комментариях и лайках помогают оценить вовлеченность аудитории и взаимодействие с ней.

Таким образом, этот датасет может быть полезным инструментом для исследования, анализа и планирования стратегий в сфере создания контента и маркетинга на платформе YouTube.

```python
data_path = "/Users/vane4ka/Downloads/Clean_Top_1000_Youtube_df - youtubers_df.csv"
data = pd.read_csv(data_path)


# Удаление запятых и преобразование в числовой формат для столбцов Likes и Visits
data['Likes'] = data['Likes'].str.replace(',', '').astype(int)
data['Visits'] = data['Visits'].str.replace(',', '').astype(int)
data['Comments'] = data['Comments'].str.replace(',', '', regex=True).str.replace('K', '').astype(float)
data['Suscribers'] = data['Suscribers'].str.replace(',', '', regex=True).str.replace('K', '').astype(float)

# Применение логарифмического масштабирования
data['Likes_log'] = np.log1p(data['Likes'])
data['Visits_log'] = np.log1p(data['Visits'])

plt.figure(figsize=(10, 6))
plt.scatter(data['Visits_log'], data['Likes_log'], alpha=0.5)
plt.title('Зависимость Лайков от Просмотров (логарифмический масштаб)')
plt.xlabel('Логарифм Просмотров')
plt.ylabel('Логарифм Лайков')
plt.grid(True)
plt.show()
```

```python
# Разделение датасета на группы по категории
category_groups = data.groupby('Categories')

# Построение корреляционной матрицы для каждой группы
for category, group in category_groups:
    if category == 'Association_football' or category == 'Christian_music' or category == 'Health' or category ==  'Independent_music' or category ==  'Mixed_martial_arts' or category == 'Pet' or category == 'Physical_fitness' or category == 'Puzzle_video_game' or category == 'Soul_music':
        continue
    print(f"Category: {category}")
    numerical_data = group.select_dtypes(include=['number'])
    correlation_matrix = numerical_data.corr()
    # Вывод корреляционной матрицы и тепловой карты для каждой категории
    plt.figure(figsize=(8, 6))
    sns.heatmap(correlation_matrix, annot=True, cmap=plt.cm.rainbow, fmt=".2f", linewidths=0.5)
    plt.title(f'Correlation Heatmap for {category}')
    plt.show()
```
