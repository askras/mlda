## Описание набора данных

1) crim: Угон преступности на душу населения по городам.
    - Этот признак может дать представление о уровне преступности в городе. Большие значения могут указывать на более высокий уровень преступности.
2) zn: Доля жилой земли, зонированной на участки более 25 000 кв. футов.
    - Позволяет оценить, какая часть земли зонирована для жилых целей.
3) indus: Доля неторговой площади на город.
    - Показывает, какую долю города занимают неторговые объекты.
4) chas: Фиктивная переменная Чарльз-Ривер (1, если участок ограничивает реку, 0 в противном случае).
    - Указывает, ограничивает ли участок реку.
5) nox: Концентрация оксидов азота (частей на 10 миллионов).
    - Уровень загрязнения воздуха оксидами азота.
6) rm: Среднее количество комнат на жилой единицу.
    - Оценка размера жилья в городе. Больше комнат обычно связано с более крупным жильем.
7) age: Доля занимаемых владельцами единиц, построенных до 1940 года.
    - Старинные дома, возможно, требуют больше внимания к обслуживанию.
8) dis: Взвешенные расстояния до пяти бостонских центров занятости.
    - Указывает на доступность к центрам занятости.
9) rad: Индекс доступности к радиальным магистралям.
    - Уровень доступности к радиальным магистралям.
10) tax: Полная ставка налога на имущество на 10 000 долларов.
    - Уровень налогов на недвижимость.
11) ptratio: Соотношение учеников и учителей по городам.
    - Оценка качества образования.
12) b: 1000(Bk - 0.63)^2, где Bk - доля афроамериканцев по городам.
    - Оценка разнообразия населения.
13) lstat: % более низкого статуса населения.
    - Оценка социального статуса.
14) medv: Среднее значение стоимости домов, занимаемых владельцами, в тысячах долларов.
    - Целевой признак, который мы обычно пытаемся предсказать.

```python
# Создаем матрицу корреляции
corr_matrix = boston_df.corr()

# Создаем тепловую карту (heatmap) для матрицы корреляции
plt.figure(figsize=(12, 10))
sns.heatmap(corr_matrix, annot=True, cmap="coolwarm", linewidths=.5)
plt.title("Матрица корреляции признаков")
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Загрузим данные из CSV файла
file_path = "dataset/BostonHousing.csv"
boston_data = pd.read_csv(file_path)

# Выведем статистическую информацию
print(boston_data.describe())

# Создаем фигуру с подграфиками
plt.figure(figsize=(18, 12))

# Перебираем числовые признаки и строим гистограммы
for i, feature in enumerate(boston_data.columns[:-1]):
    plt.subplot(3, 5, i + 1)
    sns.histplot(boston_data[feature], bins=30, kde=True)
    plt.title(f'Distribution of {feature}')

plt.tight_layout()
plt.show()
```
