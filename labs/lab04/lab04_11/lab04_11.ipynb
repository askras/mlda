Для определения, какие параметры лучше всего использовать для классификации фисташек, нужно провести анализ данных и рассмотреть, какие из параметров наиболее информативны для разделения разных видов фисташек. Однако на основе предоставленных параметров можно предположить, что следующие 5 параметров могли бы быть полезными для классификации фисташек:

1. **AREA** (площадь области, обнаруженной на изображении): Площадь может быть полезным признаком для различения разных видов фисташек, так как разные виды могут иметь разные размеры.

2. **PERIMETER** (периметр области): Периметр также может быть информативным, особенно если разные виды фисташек имеют разные формы.

3. **ECCENTRICITY** (эксцентриситет): Этот параметр описывает форму области. Разные виды фисташек могут иметь разную форму, и эксцентриситет может помочь в этом различии.

4. **SOLIDITY** (плотность): Плотность области может отражать, насколько близко фисташка круглой форме. Разные виды фисташек могут иметь разную плотность.

5. **EXTENT** (протяженность): Этот параметр может указывать на степень заполнения площади области. Разные виды фисташек могут иметь разные степени заполнения.

Это лишь предположение, и для точного выбора параметров следует провести более глубокий анализ данных и оценить их важность с использованием методов машинного обучения. Также следует учесть, что выбор параметров может зависеть от конкретной задачи классификации и доступных данных.

```python
data_source = "/Users/vane4ka/Downloads/pistachio.csv"

# Указываем только нужные столбцы при чтении CSV
selected_columns = ["AREA", "PERIMETER", "ECCENTRICITY", "SOLIDITY", "EXTENT", "Class"]
data = pd.read_csv(data_source, delimiter=',', usecols=selected_columns)

data.head(10)
```

```python
data['Class'] = data['Class'].replace({'Siit_Pistachio': 0, 'Kirmizi_Pistachio': 1})
```

```python
sns.pairplot(data, vars=["AREA", "PERIMETER", "ECCENTRICITY", "SOLIDITY", "EXTENT"]);
```

```python
sns.pairplot(data, hue='Class');
```

```python
from sklearn.neighbors import KNeighborsClassifier

X_train = data[['AREA', 'PERIMETER', 'ECCENTRICITY', 'SOLIDITY', 'EXTENT']].values
y_train = data['Class'].values

K = 29 # Количество соседей

# Создание и настройка классификатора
knn = KNeighborsClassifier(n_neighbors=K)
# построение модели классификатора (процедура обучения)
knn.fit(X_train, y_train)

# Использование классификатора
# Объявление признаков объекта
X_new = np.array([[95000, 1900, 0.8, 0.9, 0.7]])
# Получение ответа для нового объекта
target = knn.predict(X_new)
print(target)

# Использование классификатора
# Объявление признаков объекта
X_new = np.array([[75000, 1600, 0.7, 0.9, 0.7]])
# Получение ответа для нового объекта
target = knn.predict(X_new)
print(target)
```

```python
# Оценка точности классификатора с использованием методики hold-out


from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score

X_train, X_holdout, y_train, y_holdout = train_test_split(
    data[
    ['AREA',
     'PERIMETER',
     'ECCENTRICITY',
     'SOLIDITY',
     'EXTENT']], 
    data['Class'], 
    test_size=0.3, 
    random_state=17)

K = 29 # Количество соседей
knn = KNeighborsClassifier(n_neighbors=K)
knn.fit(X_train, y_train)
knn_pred = knn.predict(X_holdout)
accur = accuracy_score(y_holdout, knn_pred)
print('accuracy: ', accur)
```

```python
# Реализация процедуры выбора оптимального параметра на основе cross validation error
from sklearn.model_selection import cross_val_score

# Значения параметра K
k_list = list(range(1,50))
# Пустой список для хранения значений точности
cv_scores = []
# В цикле проходим все значения K
for K in k_list:
    knn = KNeighborsClassifier(n_neighbors=K)
    scores = cross_val_score(
        knn, 
        data.iloc[ : , 0:5 ], 
        data['Class'], 
        cv=10, 
        scoring='accuracy')
    cv_scores.append(scores.mean())

# Вычисляем ошибку (misclassification error)
MSE = [1-x for x in cv_scores]

# Строим график
plt.plot(k_list, MSE)
plt.xlabel('Количество соседей (K)');
plt.ylabel('Ошибка классификации (MSE)')
plt.show()

# Ищем минимум
k_min = min(MSE)

# Пробуем найти прочие минимумы (если их несколько)
all_k_min = []
for i in range(len(MSE)):
    if MSE[i] <= k_min:
        all_k_min.append(k_list[i])

# печатаем все K, оптимальные для модели
print('Оптимальные значения K: ', all_k_min)
```

### Контрольные вопросы

Особенности методов метрической классификации:

Метод ближайшего соседа (1NN): Классификация объекта основывается на классе его ближайшего соседа. Метод k ближайших соседей (kNN): Классификация объекта определяется большинством классов k ближайших соседей.

Принципы и этапы реализации метода kNN:

Принцип: Определение класса объекта основывается на классе его соседей. Этапы: Задание количества соседей (K). Вычисление расстояний между объектом и всеми точками обучающей выборки. Выбор K ближайших соседей. Определение класса объекта на основе большинства классов среди соседей.

Выбор количества соседей:

Количество соседей выбирается эмпирическим путем или с использованием методов оценки качества, таких как hold-out или кросс-валидация.

Метод парзеновского окна:

Это обобщение метода kNN, где вес каждого соседа зависит от расстояния до объекта. Чем дальше сосед, тем меньше его вес.

Параметры для оптимизации в методах kNN:

k (количество соседей): Оптимизируется для достижения наилучшей производительности. Метрика расстояния: Выбор метрики (евклидово расстояние, манхэттенское расстояние и т.д.) может влиять на результаты.
