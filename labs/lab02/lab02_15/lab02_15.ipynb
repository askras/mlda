![](./img/dataset-cover.jpg)

```python
import numpy as np

data_path = "datasets/house.data"
data = np.genfromtxt(data_path, delimiter=",", dtype=None)
print(data)


# Median_House_Value, Median_Income, Median_Age, Tot_Rooms, Tot_Bedrooms, Population, Households
```

```python
dt = np.dtype("int, f8, int, int, int, int, int")
data2 = np.genfromtxt(data_path, delimiter=",", dtype=dt)
print('Shape of the dataset:', data2.shape)
print('Number of cases:', data2.shape[0])
print('Dataset type:', type(data2))
print('A single row of the dataset is type of:', type(data2[0]))
print('Types of elements:', type(data2[0][1]), type(data2[0][4]))
print('Dataset slice:')
print(data2[:10])
```

##### Количество элементов набора - 20640 
##### Количество признаков - 7
##### Количество пропущенных значений - 0 

```python
import matplotlib as mpl
import matplotlib.pyplot as plt
%matplotlib inline


# Median_House_Value, Median_Income, Median_Age, Tot_Rooms, Tot_Bedrooms, Population, Households
# Data from individual columns
# median_value = [] # Средняя стоимость дома --> будем её исследовать
median_income = [] # Средняя зарплата жильцов в доме
median_age = [] # Средний возвраст дома
tot_rooms = [] # Общее количество комнат в доме
tot_bedrooms = [] # Общее количество спален в доме
population = [] # Среднее количество жильцов в доме
households = [] # Среднее количетсво семей в доме

# Traversing the entire data collection
for dot in data2:
#   media_value.append(dot[0])
    median_income.append(dot[1])
    median_age.append(dot[2])
    tot_rooms.append(dot[3])
    tot_bedrooms.append(dot[4])
    population.append(dot[5])  
    households.append(dot[6]) 
    
    

# Мы строим графики на основе проекций данных с учетом того, что cтоимость дома постепенно уменьшается
# Первая категория - от 300.000$ до 450.000$
# Вторая категория - от 150.000$ до 300.000$
# Третья категория - менее 150.000$
plt.figure(1)

Expensive, = plt.plot(tot_bedrooms[:7100], population[:7100], 'ro', label='Дорого')
Medium, = plt.plot(tot_bedrooms[7100:14200], population[7100:14200], 'g^', label='Средне')
Cheap, = plt.plot(tot_bedrooms[14200:20000], population[14200:20000], 'bs', label='Дешево')


plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
plt.xlabel('tot_bedrooms')
plt.ylabel('population')

plt.show()
```

```python
plt.figure(2)

Expensive, = plt.plot(tot_bedrooms[:7100], tot_rooms[:7100], 'ro', label='Дорого')
Medium, = plt.plot(tot_bedrooms[7100:14200], tot_rooms[7100:14200], 'g^', label='Средне')
Cheap, = plt.plot(tot_bedrooms[14200:20000], tot_rooms[14200:20000], 'bs', label='Дешево')


plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
plt.xlabel('tot_bedrooms')
plt.ylabel('tot_rooms')

plt.show()
```

```python
plt.figure(3)

Expensive, = plt.plot(population[:7100], households[:7100], 'ro', label='Дорого')
Medium, = plt.plot(population[7100:14200], households[7100:14200], 'g^', label='Средне')
Cheap, = plt.plot(population[14200:20000], households[14200:20000], 'bs', label='Дешево')


plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
plt.xlabel('population')
plt.ylabel('households')

plt.show()
```
