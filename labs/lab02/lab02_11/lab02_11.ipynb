## Описание набора данных

Набор данных ["Glass Identification"](https://archive.ics.uci.edu/dataset/42/glass+identification) представляет собой совокупность информации о различных типах стекла. Этот набор данных был предоставлен в научных целях Службой судебной экспертизы США 31 августа 1987 года. Он содержит информацию о химическом составе стекла, который определен на основе его оксидного содержания (например, Na, Fe, K и др.). Этот набор данных был использован для задачи классификации.

Характеристики набора данных:
- Мультивариативный: Набор данных содержит информацию о нескольких признаках для каждого образца стекла.
- Область исследования: Физическая наука.
- Связанные задачи: Классификация - определение типа стекла на основе его химического состава.
- Количество экземпляров: 214.
- Количество признаков: 9.

Описание признаков:
1. Id_number (Идентификатор) - уникальный идентификатор образца стекла (целочисленный).
2. RI (Показатель преломления) - непрерывный признак, представляющий собой показатель преломления стекла.
3. Na (Натрий) - непрерывный признак, представляющий процент содержания натрия в соответствующем оксиде.
4. Mg (Магний) - непрерывный признак, представляющий процент содержания магния в соответствующем оксиде.
5. Al (Алюминий) - непрерывный признак, представляющий процент содержания алюминия в соответствующем оксиде.
6. Si (Кремний) - непрерывный признак, представляющий процент содержания кремния в соответствующем оксиде.
7. K (Калий) - непрерывный признак, представляющий процент содержания калия в соответствующем оксиде.
8. Ca (Кальций) - непрерывный признак, представляющий процент содержания кальция в соответствующем оксиде.
9. Ba (Барий) - непрерывный признак, представляющий процент содержания бария в соответствующем оксиде.
10. Fe (Железо) - непрерывный признак, представляющий процент содержания железа в соответствующем оксиде.

Назначение набора данных:
Набор данных "Glass Identification" был создан для классификации различных типов стекла на основе их химического состава. Эта задача может быть полезной, например, для криминалистов, чтобы определить тип стекла, найденного на месте преступления, и использовать его в качестве улик.

Возможные модели, которые можно построить на основе данного набора данных:
1. Модель классификации: Можно построить модель машинного обучения для определения типа стекла на основе его химического состава. Для этого можно использовать различные алгоритмы классификации, такие как логистическая регрессия, метод ближайших соседей (k-NN), дискриминантный анализ и другие.

2. Анализ природы стекла: Набор данных также может быть использован для анализа свойств и характеристик стекла в зависимости от его химического состава.

3. Визуализация данных: Можно создать графики и визуализации для лучшего понимания химического состава стекла и его свойств.

Этот набор данных может быть полезным для обучения и оценки производительности различных моделей классификации и проведения исследований в области химии стекла.

```python
import numpy as np
import matplotlib.pyplot as plt

data_path = "/Users/vane4ka/Desktop/glass+identification/glass.data"

data = np.genfromtxt(data_path, delimiter=",")

print("Тип данных:", type(data))
print("Форма данных:", data.shape)
print(data[-4:])
```
